```python
# ============================================================
# CALIGO FORMAL FIGURE NOTEBOOK — FULL CLEAN VERSION
# Target notebook:
#   /Users/svc553/Documents/Caligo/examples/notebooks/figures.ipynb
#
# What this does:
#   1. Builds the GJ1214b Caligo case from examples/run_gj1214b.py
#      but stops before the old case.haze(fake_out) call.
#   2. Runs photochem.
#   3. Runs a small haze-source comparison grid.
#   4. Explicitly builds source_info with case.build_haze_source().
#   5. Runs full microphysics.
#   6. Saves every run to /Users/svc553/Documents/Caligo/output/...
#   7. Makes paper-style figures:
#        - P-T and Kzz
#        - source profile
#        - final haze mass-density profile
#        - pressure-radius haze mass map
#        - radius versus pressure
#        - timescales versus pressure
#        - settling speed versus pressure
#        - size distributions
#        - flux-budget history
#        - direct-Mie haze opacity spectra
#        - wavelength-pressure opacity maps
#        - source-flux summary plots
# ============================================================


# ============================================================
# 0. Imports and global paths
# ============================================================

from pathlib import Path
import os
import sys
import time
import json
import traceback
import re

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm


PROJECT_ROOT = Path("/Users/svc553/Documents/Caligo").resolve()
EXAMPLE_SCRIPT = PROJECT_ROOT / "examples" / "run_gj1214b.py"

# New clean output folder. Keeping this separate avoids old failed output pollution.
OUTPUT_ROOT = PROJECT_ROOT / "output" / "formal_figures_flux_comparison_v2"
OUTPUT_ROOT.mkdir(parents=True, exist_ok=True)

REFRIND_DIR = Path("/Users/svc553/virga/virga/lab_ref_mieff").resolve()
REFRIND_BASENAME = "ExoHaze1000xSolar_300K"

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

os.chdir(PROJECT_ROOT)

print("PROJECT_ROOT:", PROJECT_ROOT)
print("EXAMPLE_SCRIPT:", EXAMPLE_SCRIPT)
print("OUTPUT_ROOT:", OUTPUT_ROOT)


# ============================================================
# 1. Import Caligo and optional Mie package
# ============================================================

import caligo
from caligo import diagnostics as diag

try:
    import miepython
    HAS_MIEPYTHON = True
    print("miepython imported successfully.")
except Exception as exc:
    HAS_MIEPYTHON = False
    print("miepython not available. Direct-Mie opacity plots will be skipped.")
    print("Reason:", repr(exc))


# ============================================================
# 2. Plot/save helpers
# ============================================================

plt.rcParams.update({
    "figure.figsize": (7.5, 5.0),
    "axes.grid": True,
    "grid.alpha": 0.28,
    "font.size": 11,
})


def ensure_dir(path):
    path = Path(path)
    path.mkdir(parents=True, exist_ok=True)
    return path


def savefig(fig, path):
    path = Path(path)
    path.parent.mkdir(parents=True, exist_ok=True)
    fig.savefig(path, dpi=220, bbox_inches="tight")
    print("saved:", path)


def safe_float(x, default=np.nan):
    try:
        return float(x)
    except Exception:
        return default


def get_result_array(result, key, default=None):
    if isinstance(result, dict) and key in result:
        return np.asarray(result[key])

    if isinstance(result, dict) and "ctx" in result and isinstance(result["ctx"], dict):
        if key in result["ctx"]:
            return np.asarray(result["ctx"][key])

    return default


def get_result_value(result, key, default=None):
    if isinstance(result, dict) and key in result:
        return result[key]

    if isinstance(result, dict) and "ctx" in result and isinstance(result["ctx"], dict):
        if key in result["ctx"]:
            return result["ctx"][key]

    return default


def pressure_sort_top_to_bottom(Pbar, *arrays):
    Pbar = np.asarray(Pbar)
    idx = np.argsort(Pbar)
    out = [Pbar[idx]]

    for arr in arrays:
        arr = np.asarray(arr)
        if arr.shape[0] == len(Pbar):
            out.append(arr[idx, ...])
        else:
            out.append(arr)

    return out


def write_json(path, obj):
    path = Path(path)
    path.parent.mkdir(parents=True, exist_ok=True)
    with open(path, "w") as f:
        json.dump(obj, f, indent=2)


# ============================================================
# 3. Build base case from example setup only
# ============================================================

def truncate_example_before_old_haze_call(script_path):
    """
    Executes examples/run_gj1214b.py only up to the setup part.

    This avoids running the old:
        haze = case.haze(fake_out)

    call inside the example script.
    """
    script_path = Path(script_path)
    text = script_path.read_text()
    lines = text.splitlines()

    stop_patterns = [
        r"^\s*haze\s*=\s*case\.haze\s*\(",
        r"^\s*case\.haze\s*\(",
        r"^\s*case\.run_microphysics\s*\(",
        r"^\s*case\.plot\s*\(",
        r"^\s*case\.summary\s*\(",
    ]

    stop_idx = len(lines)

    for i, line in enumerate(lines):
        for pat in stop_patterns:
            if re.search(pat, line):
                stop_idx = min(stop_idx, i)
                break

    setup_lines = lines[:stop_idx]
    setup_code = "\n".join(setup_lines)

    print("Original example lines:", len(lines))
    print("Executing setup lines:", len(setup_lines))

    if stop_idx < len(lines):
        print("Stopped before line:")
        print(lines[stop_idx])

    return setup_code


def find_case_in_namespace(ns):
    if "case" in ns:
        return ns["case"]

    candidates = []

    for key, value in ns.items():
        if key.startswith("_"):
            continue

        if hasattr(value, "run_microphysics") and hasattr(value, "haze_source"):
            candidates.append((key, value))

    if len(candidates) == 1:
        print("Found case-like object:", candidates[0][0])
        return candidates[0][1]

    if len(candidates) > 1:
        print("Multiple case-like objects found:")
        for key, _ in candidates:
            print(" ", key)
        return candidates[0][1]

    return None


def build_base_case_from_example_setup():
    if not EXAMPLE_SCRIPT.exists():
        raise FileNotFoundError(f"Could not find {EXAMPLE_SCRIPT}")

    setup_code = truncate_example_before_old_haze_call(EXAMPLE_SCRIPT)

    ns = {
        "__name__": "__caligo_setup_only__",
        "__file__": str(EXAMPLE_SCRIPT),
    }

    exec(compile(setup_code, str(EXAMPLE_SCRIPT), "exec"), ns)

    case = find_case_in_namespace(ns)

    if case is None:
        raise RuntimeError(
            "Could not find a Caligo case object after executing the setup "
            "part of examples/run_gj1214b.py."
        )

    fake_out = ns.get("fake_out", None)
    pt = ns.get("pt", None)

    print("\nBase case built from example setup.")
    print("case type:", type(case))
    print("fake_out found:", fake_out is not None)
    print("pt found:", pt is not None)

    return case, fake_out, pt, ns


base_case, base_fake_out, base_pt, setup_namespace = build_base_case_from_example_setup()


# ============================================================
# 4. Run photochem
# ============================================================

def ensure_photochem_ready(case_obj, find_steady_state=True):
    photochem_result = getattr(case_obj, "_photochem_result", None)

    if photochem_result is not None:
        print("Photochem result already exists. Skipping run_photochem.")
        return photochem_result

    print("\nRunning case.run_photochem(find_steady_state=True)...")
    t0 = time.time()

    pc, Pbar_pc, z_pc, T_pc = case_obj.run_photochem(
        find_steady_state=find_steady_state
    )

    runtime_s = time.time() - t0
    photochem_result = getattr(case_obj, "_photochem_result", None)

    print("\nPhotochem finished.")
    print("runtime [min]:", runtime_s / 60.0)
    print("Pbar_pc range:", np.nanmin(Pbar_pc), np.nanmax(Pbar_pc))
    print("z_pc range:", np.nanmin(z_pc), np.nanmax(z_pc))

    return photochem_result


photochem_result = ensure_photochem_ready(base_case, find_steady_state=True)


# ============================================================
# 5. Science settings and cases
# ============================================================

SOLVER_SETTINGS = {
    "dt_init": 1.0e2,
    "dt_growth": 1.03,
    "dt_max": 300.0,
    "chunk_time": 1.0e5,
    "max_chunks": 200,
    "max_step_retries": 10,
    "retry_shrink": 0.5,
    "ss_tol_storage": 0.08,
    "ss_tol_balance": 0.25,
    "ss_tol_profile": 0.04,
    "ss_tol_peak_dex": 0.25,
    "ss_min_stable_chunks": 3,
}

# By default, run two cases.
# Add the 3e-13 case if you want the near-optically-thick comparison.
CASES = [
    {
        "label": "flux1e-13_peak1e-7",
        "target_haze_flux": 1.0e-13,
        "gauss_peak_bar": 1.0e-7,
        "gauss_sigma_dex": 0.50,
    },
    {
        "label": "flux2e-13_peak1e-7",
        "target_haze_flux": 2.0e-13,
        "gauss_peak_bar": 1.0e-7,
        "gauss_sigma_dex": 0.50,
    },

    # Uncomment if you want the stronger case too.
    # This may take ~8–10 hours by itself.
    # {
    #     "label": "flux3e-13_peak1e-7",
    #     "target_haze_flux": 3.0e-13,
    #     "gauss_peak_bar": 1.0e-7,
    #     "gauss_sigma_dex": 0.50,
    # },
]

RERUN_EXISTING = False

catalog_path = OUTPUT_ROOT / "run_catalog.csv"

print("\nCases to run:")
for c in CASES:
    print(c)


def reset_caligo_runtime_cache(case_obj):
    """
    Clear previous haze/microphysics products.

    Critical:
      _source_info is cleared so each flux case gets its own source.
    """
    cache_attrs = [
        "_ctx",
        "_haze_out",
        "_microphysics_result",
        "_result",
        "_last_result",
        "_last_haze",
        "_source_info",
    ]

    for attr in cache_attrs:
        if hasattr(case_obj, attr):
            try:
                setattr(case_obj, attr, None)
            except Exception:
                pass


def attach_science_settings(
    case_obj,
    target_haze_flux,
    gauss_peak_bar=1.0e-7,
    gauss_sigma_dex=0.50,
):
    """
    Attach the haze, particle, aggregate, coagulation, transport,
    sink, and solver settings.
    """
    case_obj.haze_source(
        vertical_mode="gaussian",
        coupling_mode="split_top_flux",
        parents=["HCN", "C2H2"],
        yield_model="carbon_mass_only",
        yield_haze=0.10,
        use_fixed_flux=True,
        target_haze_flux=target_haze_flux,
        source_flux_scale=1.0,
        gauss_peak_bar=gauss_peak_bar,
        gauss_sigma_dex=gauss_sigma_dex,
        bin_mode="single_bin",
        bin_index=0,
        r0_nm=2.0,
        sigma_dex=0.20,
        bin_weights=None,
    )

    case_obj.particles(
        n_bin=40,
        rho_haze=1.0,
        r_monomer_nm=1.0,
        bin_mass_ratio=2.0,
    )

    case_obj.aggregates(
        use=True,
        df_fixed=2.0,
        r_primary_nm=10.0,
        k0=1.0,
        pressure_dependent=True,
        transition_p_bar=1.0e-6,
        transition_width_dex=0.5,
        use_agg_in_settling=True,
        use_agg_in_coag=True,
        collision_radius_cap_factor=None,
    )

    case_obj.coagulation(
        use=True,
        brownian=True,
        differential_settling=False,
        kernel_scale=1.0,
        charge_suppression=True,
        charge_density_e_per_um=15.0,
        substep_frac=0.03,
        max_substeps=10000,
        fixed_point_iters=4,
        skip_ntot=1e2,
        skip_mass_dens=1e-30,
        skip_dtl=1e-3,
        active_rel=1e-8,
        mass_err_warn=1e-6,
        top_overflow_mode="mass_conserve",
        retry_on_coag_substep_cap=True,
    )

    case_obj.transport(
        settling=True,
        settling_scale=1.0,
        bottom_bc="deposition_velocity",
        sedimentation_cfl=0.60,
        max_sedimentation_substeps=10000,
        transport_active_frac=1e-24,
    )

    case_obj.sinks(
        thermal_model="finite_rate_smooth_T",
        therm_destroy_t_on_k=650.0,
        therm_destroy_width_k=100.0,
        therm_destroy_timescale=1.0e5,
        enable_recycling_to_gas=False,
        recycled_gas_label="passive_recycled_haze_mass",
        recycle_gas_diffusion=False,
        use_deep_recycling=False,
        deep_recycling_p_on_bar=1e-3,
        deep_recycling_width_dex=0.5,
        deep_recycling_timescale=1.0e6,
        deep_recycling_to_gas=False,
    )

    case_obj.solver(**SOLVER_SETTINGS)


def ensure_source_info_built(case_obj):
    """
    Build source_info for the current haze_source settings.

    This is the missing step that caused the previous failure.
    """
    photochem_result = getattr(case_obj, "_photochem_result", None)

    if photochem_result is None:
        raise RuntimeError(
            "No photochem result exists. Run case.run_photochem() before "
            "building the haze source."
        )

    print("\nBuilding haze source for current settings...", flush=True)

    if not hasattr(case_obj, "build_haze_source"):
        raise AttributeError("case_obj has no build_haze_source method.")

    # Your code's error message says to call case.build_haze_source()
    # with no args. Keep a keyword fallback in case the local signature changed.
    try:
        source_info = case_obj.build_haze_source()
    except TypeError:
        source_info = case_obj.build_haze_source(
            photochem_result=photochem_result
        )

    case_obj._source_info = source_info

    print("Haze source built.", flush=True)

    if isinstance(source_info, dict):
        print("source_info keys:", list(source_info.keys()))

        for key in [
            "F_src",
            "F_raw",
            "F_vol_source",
            "F_top_source",
            "source_scale",
            "peak_pressure_bar",
            "precursors",
        ]:
            if key in source_info:
                print(f"  {key}: {source_info[key]}", flush=True)

    return source_info


# ============================================================
# 6. Direct-Mie opacity tools
# ============================================================

def find_refrind_file(refrind_dir, basename):
    refrind_dir = Path(refrind_dir)

    candidates = []

    for pattern in [
        basename + ".refrind",
        basename + "*.refrind",
        "*" + basename + "*.refrind",
    ]:
        candidates.extend(sorted(refrind_dir.glob(pattern)))

    candidates = [p for p in candidates if p.is_file()]

    if len(candidates) == 0:
        raise FileNotFoundError(
            f"No .refrind file found for {basename} in {refrind_dir}"
        )

    exact = refrind_dir / f"{basename}.refrind"

    if exact.exists():
        return exact

    return candidates[0]


def read_numeric_rows_any_width(path):
    rows = []
    header = []

    with open(path, "r", errors="ignore") as f:
        for line in f:
            raw = line.strip()

            if raw == "" or raw.startswith("#") or raw.startswith("!"):
                header.append(raw)
                continue

            parts = raw.replace(",", " ").split()

            vals = []
            ok = True

            for p in parts:
                try:
                    vals.append(float(p))
                except Exception:
                    ok = False
                    break

            if ok and len(vals) >= 3:
                rows.append(vals)
            else:
                header.append(raw)

    if len(rows) == 0:
        raise ValueError(f"No numeric rows found in {path}")

    max_len = max(len(r) for r in rows)
    rows = [r for r in rows if len(r) == max_len]

    arr = np.asarray(rows, dtype=float)
    return arr, header


def convert_wave_if_needed(wave):
    wave = np.asarray(wave, dtype=float)
    wmax = np.nanmax(wave)

    if wmax > 1e3:
        return wave / 1e4, "angstrom_to_micron"
    elif wmax < 1e-2:
        return wave * 1e4, "cm_to_micron"
    else:
        return wave, "micron"


def score_refrind_candidate(wave, n_real, k_imag):
    wave = np.asarray(wave, dtype=float)
    n_real = np.asarray(n_real, dtype=float)
    k_imag = np.asarray(k_imag, dtype=float)

    finite = (
        np.isfinite(wave)
        & np.isfinite(n_real)
        & np.isfinite(k_imag)
        & (wave > 0)
        & (n_real > 0)
        & (k_imag >= 0)
    )

    if np.sum(finite) < 10:
        return -np.inf

    wave = wave[finite]
    n_real = n_real[finite]
    k_imag = k_imag[finite]

    score = 0.0

    if wave.min() < 1.0 and wave.max() > 5.0:
        score += 5.0
    if wave.min() < 0.5 and wave.max() > 10.0:
        score += 5.0

    n_med = np.nanmedian(n_real)
    n_min = np.nanmin(n_real)
    n_max = np.nanmax(n_real)

    if 1.0 <= n_med <= 3.0:
        score += 10.0
    if 0.5 <= n_min and n_max <= 5.0:
        score += 5.0

    k_med = np.nanmedian(k_imag)
    k_max = np.nanmax(k_imag)

    if 0.0 <= k_med <= 2.0:
        score += 5.0
    if k_max <= 10.0:
        score += 2.0

    if n_max > 10.0:
        score -= 20.0
    if n_min < 0.1:
        score -= 5.0

    return score


def read_refrind_corrected(path):
    arr, header = read_numeric_rows_any_width(path)

    ncol = arr.shape[1]
    candidates = []

    wave0, unit0 = convert_wave_if_needed(arr[:, 0])
    candidates.append({
        "name": "col0_wave_col1_n_col2_k",
        "wave_um": wave0,
        "n_real": arr[:, 1],
        "k_imag": np.abs(arr[:, 2]),
        "unit": unit0,
    })

    if ncol >= 4:
        wave1, unit1 = convert_wave_if_needed(arr[:, 1])
        candidates.append({
            "name": "col0_index_col1_wave_col2_n_col3_k",
            "wave_um": wave1,
            "n_real": arr[:, 2],
            "k_imag": np.abs(arr[:, 3]),
            "unit": unit1,
        })

    for c in candidates:
        c["score"] = score_refrind_candidate(
            c["wave_um"],
            c["n_real"],
            c["k_imag"],
        )

    candidates = sorted(candidates, key=lambda c: c["score"], reverse=True)
    best = candidates[0]

    print("\nRefrind column-map candidates")
    print("-----------------------------")

    for c in candidates:
        print(c["name"], "score =", c["score"], "unit =", c["unit"])
        print("  wave range:", np.nanmin(c["wave_um"]), np.nanmax(c["wave_um"]))
        print("  n range:", np.nanmin(c["n_real"]), np.nanmax(c["n_real"]))
        print("  k range:", np.nanmin(c["k_imag"]), np.nanmax(c["k_imag"]))

    print("\nSelected refrind mapping:", best["name"])

    df = pd.DataFrame({
        "wave_um": best["wave_um"],
        "n_real": best["n_real"],
        "k_imag": best["k_imag"],
    })

    df = df.replace([np.inf, -np.inf], np.nan)
    df = df.dropna(subset=["wave_um", "n_real", "k_imag"])
    df = df[df["wave_um"] > 0]
    df = df[df["n_real"] > 0]
    df = df[df["k_imag"] >= 0]

    df = (
        df.groupby("wave_um", as_index=False)
        .agg({"n_real": "mean", "k_imag": "mean"})
        .sort_values("wave_um")
        .reset_index(drop=True)
    )

    return df, header, best


def mie_efficiencies_scalar(m_complex, radius_um, wave_um_value):
    radius_um = float(radius_um)
    wave_um_value = float(wave_um_value)

    if radius_um <= 0 or wave_um_value <= 0:
        return 0.0, 0.0, 0.0, 0.0

    diameter_um = 2.0 * radius_um
    x = 2.0 * np.pi * radius_um / wave_um_value

    if hasattr(miepython, "efficiencies"):
        qext, qsca, qback, g = miepython.efficiencies(
            m_complex,
            diameter_um,
            wave_um_value,
        )[:4]
    elif hasattr(miepython, "mie"):
        qext, qsca, qback, g = miepython.mie(m_complex, x)
    elif hasattr(miepython, "efficiencies_mx"):
        qext, qsca, qback, g = miepython.efficiencies_mx(m_complex, x)
    else:
        raise RuntimeError("No compatible miepython API found.")

    qext = max(float(np.real(qext)), 0.0)
    qsca = max(float(np.real(qsca)), 0.0)

    if qsca > qext:
        qsca = qext

    qabs = max(qext - qsca, 0.0)

    g = float(np.real(g))
    if not np.isfinite(g):
        g = 0.0

    return qext, qsca, qabs, g


def compute_direct_mie_opacity(result, run_dir, label, n_wave=300):
    if not HAS_MIEPYTHON:
        print("Skipping direct-Mie opacity because miepython is unavailable.")
        return None

    try:
        refrind_path = find_refrind_file(REFRIND_DIR, REFRIND_BASENAME)
        refr, refr_header, refr_mapping = read_refrind_corrected(refrind_path)
    except Exception as exc:
        print("Skipping direct-Mie opacity. Could not read refrind.")
        print("Reason:", repr(exc))
        return None

    optics_dir = ensure_dir(run_dir / "direct_mie")

    wave_min = max(0.400008, float(refr["wave_um"].min()))
    wave_max = min(20.0, float(refr["wave_um"].max()))
    wave_um = np.logspace(np.log10(wave_min), np.log10(wave_max), n_wave)

    n_real_wave = np.interp(
        wave_um,
        refr["wave_um"].values,
        refr["n_real"].values,
        left=refr["n_real"].values[0],
        right=refr["n_real"].values[-1],
    )

    k_positive = np.maximum(refr["k_imag"].values, 1e-300)

    k_imag_wave = 10.0 ** np.interp(
        np.log10(wave_um),
        np.log10(refr["wave_um"].values),
        np.log10(k_positive),
        left=np.log10(k_positive[0]),
        right=np.log10(k_positive[-1]),
    )

    # miepython convention for absorbing materials.
    m_complex_wave = n_real_wave - 1j * k_imag_wave

    Pbar = get_result_array(result, "Pbar")
    T = get_result_array(result, "T")
    dz = get_result_array(result, "dz")
    n_cm3 = get_result_array(result, "n")
    r_nm = get_result_array(result, "r_compact_nm")
    H = get_result_array(result, "H")

    if Pbar is None or dz is None or n_cm3 is None or r_nm is None:
        print("Skipping direct-Mie opacity because required arrays are missing.")
        return None

    if H is None:
        H = np.ones_like(Pbar) * np.nanmedian(dz)

    r_planet = None

    for key in ["r_planet", "planet_radius", "Rp", "radius"]:
        val = get_result_value(result, key, None)
        if val is not None:
            try:
                r_planet = float(val)
                break
            except Exception:
                pass

    if r_planet is None:
        # GJ 1214b-like fallback used in your previous runs.
        r_planet = 1.74297496e9

    nz, nbin = n_cm3.shape

    r_um = r_nm * 1e-3
    r_cm = r_nm * 1e-7

    qext_bins = np.zeros((nbin, wave_um.size))
    qsca_bins = np.zeros((nbin, wave_um.size))
    qabs_bins = np.zeros((nbin, wave_um.size))
    g_bins = np.zeros((nbin, wave_um.size))

    print(f"\nComputing direct-Mie opacity for {label}")

    for i in range(nbin):
        if i % 10 == 0 or i == nbin - 1:
            print(f"  bin {i+1}/{nbin}: r = {r_nm[i]:.3e} nm")

        for j in range(wave_um.size):
            qext, qsca, qabs, g = mie_efficiencies_scalar(
                m_complex_wave[j],
                r_um[i],
                wave_um[j],
            )

            qext_bins[i, j] = qext
            qsca_bins[i, j] = qsca
            qabs_bins[i, j] = qabs
            g_bins[i, j] = g

    dtau_ext_layer = np.zeros((nz, wave_um.size))
    dtau_abs_layer = np.zeros((nz, wave_um.size))
    dtau_sca_layer = np.zeros((nz, wave_um.size))

    tau_ext_by_bin = np.zeros((nbin, wave_um.size))
    tau_abs_by_bin = np.zeros((nbin, wave_um.size))
    tau_sca_by_bin = np.zeros((nbin, wave_um.size))

    for i in range(nbin):
        area = np.pi * r_cm[i] ** 2
        col_num_i = np.sum(n_cm3[:, i] * dz)

        tau_ext_by_bin[i, :] = col_num_i * area * qext_bins[i, :]
        tau_abs_by_bin[i, :] = col_num_i * area * qabs_bins[i, :]
        tau_sca_by_bin[i, :] = col_num_i * area * qsca_bins[i, :]

        for k in range(nz):
            layer_num = n_cm3[k, i] * dz[k]
            dtau_ext_layer[k, :] += layer_num * area * qext_bins[i, :]
            dtau_abs_layer[k, :] += layer_num * area * qabs_bins[i, :]
            dtau_sca_layer[k, :] += layer_num * area * qsca_bins[i, :]

    vertical_tau_ext = np.sum(dtau_ext_layer, axis=0)
    vertical_tau_abs = np.sum(dtau_abs_layer, axis=0)
    vertical_tau_sca = np.sum(dtau_sca_layer, axis=0)

    H_safe = np.maximum(np.asarray(H, dtype=float), 1e-300)
    slant_factor_layer = np.sqrt(2.0 * np.pi * r_planet / H_safe)

    slant_dtau_ext_layer = dtau_ext_layer * slant_factor_layer[:, None]
    slant_dtau_abs_layer = dtau_abs_layer * slant_factor_layer[:, None]
    slant_dtau_sca_layer = dtau_sca_layer * slant_factor_layer[:, None]

    slant_tau_ext = np.sum(slant_dtau_ext_layer, axis=0)
    slant_tau_abs = np.sum(slant_dtau_abs_layer, axis=0)
    slant_tau_sca = np.sum(slant_dtau_sca_layer, axis=0)

    p_eff = np.sum(dtau_ext_layer * Pbar[:, None], axis=0) / np.maximum(
        np.sum(dtau_ext_layer, axis=0),
        1e-300,
    )

    tau_ref = np.interp(1.0, wave_um, slant_tau_ext)
    relative_slant_proxy = np.log1p(slant_tau_ext) - np.log1p(tau_ref)
    transmission_strength = 1.0 - np.exp(-slant_tau_ext)

    tau_df = pd.DataFrame({
        "wave_um": wave_um,
        "n_real": n_real_wave,
        "k_imag": k_imag_wave,
        "vertical_tau_ext": vertical_tau_ext,
        "vertical_tau_abs": vertical_tau_abs,
        "vertical_tau_sca": vertical_tau_sca,
        "slant_tau_ext": slant_tau_ext,
        "slant_tau_abs": slant_tau_abs,
        "slant_tau_sca": slant_tau_sca,
        "p_eff_bar": p_eff,
        "relative_slant_proxy": relative_slant_proxy,
        "transmission_strength": transmission_strength,
    })

    tau_csv = optics_dir / f"{label}_direct_mie_tau.csv"
    tau_npz = optics_dir / f"{label}_direct_mie_tau.npz"
    layer_npz = optics_dir / f"{label}_direct_mie_layer_dtau.npz"
    qbin_npz = optics_dir / f"{label}_direct_mie_q_by_bin.npz"

    tau_df.to_csv(tau_csv, index=False)

    np.savez_compressed(
        tau_npz,
        wave_um=wave_um,
        vertical_tau_ext=vertical_tau_ext,
        vertical_tau_abs=vertical_tau_abs,
        vertical_tau_sca=vertical_tau_sca,
        slant_tau_ext=slant_tau_ext,
        slant_tau_abs=slant_tau_abs,
        slant_tau_sca=slant_tau_sca,
        p_eff_bar=p_eff,
        relative_slant_proxy=relative_slant_proxy,
        transmission_strength=transmission_strength,
        tau_ext_by_bin=tau_ext_by_bin,
        tau_abs_by_bin=tau_abs_by_bin,
        tau_sca_by_bin=tau_sca_by_bin,
        r_compact_nm=r_nm,
    )

    np.savez_compressed(
        layer_npz,
        wave_um=wave_um,
        Pbar=Pbar,
        T=T,
        dz=dz,
        dtau_ext_layer=dtau_ext_layer,
        dtau_abs_layer=dtau_abs_layer,
        dtau_sca_layer=dtau_sca_layer,
        slant_dtau_ext_layer=slant_dtau_ext_layer,
        slant_dtau_abs_layer=slant_dtau_abs_layer,
        slant_dtau_sca_layer=slant_dtau_sca_layer,
        slant_factor_layer=slant_factor_layer,
    )

    np.savez_compressed(
        qbin_npz,
        wave_um=wave_um,
        r_compact_nm=r_nm,
        qext_bins=qext_bins,
        qabs_bins=qabs_bins,
        qsca_bins=qsca_bins,
        g_bins=g_bins,
    )

    summary = {
        "tau_csv": str(tau_csv),
        "tau_npz": str(tau_npz),
        "layer_npz": str(layer_npz),
        "qbin_npz": str(qbin_npz),
        "refrind_path": str(refrind_path),
        "refrind_mapping": refr_mapping["name"],
        "max_vertical_tau_ext": float(np.nanmax(vertical_tau_ext)),
        "max_slant_tau_ext": float(np.nanmax(slant_tau_ext)),
        "max_transmission_strength": float(np.nanmax(transmission_strength)),
        "wave_um_at_max_tau": float(wave_um[np.nanargmax(slant_tau_ext)]),
        "p_eff_at_max_tau_bar": float(p_eff[np.nanargmax(slant_tau_ext)]),
    }

    write_json(optics_dir / f"{label}_direct_mie_summary.json", summary)

    print("Direct-Mie max slant tau:", summary["max_slant_tau_ext"])
    print("Direct-Mie max transmission strength:", summary["max_transmission_strength"])

    return {
        "summary": summary,
        "tau_df": tau_df,
        "wave_um": wave_um,
        "dtau_ext_layer": dtau_ext_layer,
        "slant_dtau_ext_layer": slant_dtau_ext_layer,
        "qext_bins": qext_bins,
        "qabs_bins": qabs_bins,
        "qsca_bins": qsca_bins,
        "g_bins": g_bins,
        "tau_ext_by_bin": tau_ext_by_bin,
        "tau_abs_by_bin": tau_abs_by_bin,
        "tau_sca_by_bin": tau_sca_by_bin,
    }


# ============================================================
# 7. Run one case
# ============================================================

def run_one_case(case_obj, variant):
    label = variant["label"]
    run_dir = ensure_dir(OUTPUT_ROOT / label)

    result_npz = run_dir / f"{label}_result.npz"
    catalog_json = run_dir / f"{label}_catalog_row.json"

    if result_npz.exists() and catalog_json.exists() and not RERUN_EXISTING:
        print("\nSkipping existing successful run:", label)
        with open(catalog_json, "r") as f:
            row = json.load(f)
        return row

    print("\n" + "=" * 80)
    print("RUNNING CASE:", label)
    print("=" * 80)
    print(variant)

    t0 = time.time()

    reset_caligo_runtime_cache(case_obj)

    attach_science_settings(
        case_obj,
        target_haze_flux=variant["target_haze_flux"],
        gauss_peak_bar=variant["gauss_peak_bar"],
        gauss_sigma_dex=variant["gauss_sigma_dex"],
    )

    settings_path = run_dir / f"{label}_settings.json"
    write_json(settings_path, variant)

    try:
        source_info = ensure_source_info_built(case_obj)

        source_info_path = run_dir / f"{label}_source_info_keys.txt"
        with open(source_info_path, "w") as f:
            f.write("source_info keys\n")
            f.write("================\n")
            if isinstance(source_info, dict):
                for k in source_info.keys():
                    f.write(str(k) + "\n")
            else:
                f.write(f"source_info type: {type(source_info)}\n")

        result = case_obj.run_microphysics(
            prepare_context=True,
            dt_init=SOLVER_SETTINGS["dt_init"],
            dt_growth=SOLVER_SETTINGS["dt_growth"],
            dt_max=SOLVER_SETTINGS["dt_max"],
            chunk_time=SOLVER_SETTINGS["chunk_time"],
            max_chunks=SOLVER_SETTINGS["max_chunks"],
            max_step_retries=SOLVER_SETTINGS["max_step_retries"],
            retry_shrink=SOLVER_SETTINGS["retry_shrink"],
            ss_tol_storage=SOLVER_SETTINGS["ss_tol_storage"],
            ss_tol_balance=SOLVER_SETTINGS["ss_tol_balance"],
            ss_tol_profile=SOLVER_SETTINGS["ss_tol_profile"],
            ss_tol_peak_dex=SOLVER_SETTINGS["ss_tol_peak_dex"],
            ss_min_stable_chunks=SOLVER_SETTINGS["ss_min_stable_chunks"],
            verbose=True,
        )

        runtime_s = time.time() - t0

        history_csv = run_dir / f"{label}_history.csv"
        summary_txt = run_dir / f"{label}_summary.txt"

        diag.save_result_npz(result, result_npz)
        diag.save_history_csv(result=result, filename=history_csv)
        diag.save_summary_txt(case_obj, result=result, filename=summary_txt)

        optics = compute_direct_mie_opacity(result, run_dir, label, n_wave=300)

        last = result.get("last_chunk", {})

        row = {
            "label": label,
            "status": "success",
            "runtime_min": runtime_s / 60.0,
            "converged": bool(result.get("converged", False)),
            "target_haze_flux": variant["target_haze_flux"],
            "gauss_peak_bar": variant["gauss_peak_bar"],
            "gauss_sigma_dex": variant["gauss_sigma_dex"],
            "column_mass_g_cm2": safe_float(result.get("column_mass")),
            "peak_pressure_bar": safe_float(result.get("peak_pressure")),
            "number_weighted_radius_nm": safe_float(result.get("number_weighted_radius_nm")),
            "mass_weighted_radius_nm": safe_float(result.get("mass_weighted_radius_nm")),
            "t_final_days": safe_float(result.get("t_final")) / 86400.0,
            "F_source": safe_float(last.get("F_source")),
            "F_sink": safe_float(last.get("F_sink")),
            "F_storage": safe_float(last.get("F_storage")),
            "storage_ratio": safe_float(last.get("storage_ratio")),
            "balance_ratio": safe_float(last.get("balance_ratio")),
            "profile_change": safe_float(last.get("profile_change")),
            "run_dir": str(run_dir),
            "result_npz": str(result_npz),
            "history_csv": str(history_csv),
            "summary_txt": str(summary_txt),
            "settings_path": str(settings_path),
            "source_info_path": str(source_info_path),
            "error": "",
        }

        if optics is not None:
            row.update({
                "tau_csv": optics["summary"]["tau_csv"],
                "tau_npz": optics["summary"]["tau_npz"],
                "layer_npz": optics["summary"]["layer_npz"],
                "qbin_npz": optics["summary"]["qbin_npz"],
                "max_vertical_tau_ext": optics["summary"]["max_vertical_tau_ext"],
                "max_slant_tau_ext": optics["summary"]["max_slant_tau_ext"],
                "max_transmission_strength": optics["summary"]["max_transmission_strength"],
                "wave_um_at_max_tau": optics["summary"]["wave_um_at_max_tau"],
                "p_eff_at_max_tau_bar": optics["summary"]["p_eff_at_max_tau_bar"],
            })

        write_json(catalog_json, row)

        print("\nCASE COMPLETE:", label)
        print("runtime [min]:", row["runtime_min"])
        print("converged:", row["converged"])
        print("column mass:", row["column_mass_g_cm2"])
        print("max slant tau:", row.get("max_slant_tau_ext", np.nan))

        return row

    except Exception as exc:
        runtime_s = time.time() - t0

        error_path = run_dir / f"{label}_error.txt"
        with open(error_path, "w") as f:
            f.write(traceback.format_exc())

        row = {
            "label": label,
            "status": "failed",
            "runtime_min": runtime_s / 60.0,
            "target_haze_flux": variant["target_haze_flux"],
            "gauss_peak_bar": variant["gauss_peak_bar"],
            "gauss_sigma_dex": variant["gauss_sigma_dex"],
            "run_dir": str(run_dir),
            "error": repr(exc),
            "error_path": str(error_path),
        }

        write_json(catalog_json, row)

        print("\nCASE FAILED:", label)
        print("error:", repr(exc))
        print("error_path:", error_path)

        return row


# ============================================================
# 8. Run cases
# ============================================================

catalog_rows = []

for variant in CASES:
    row = run_one_case(base_case, variant)
    catalog_rows.append(row)

    pd.DataFrame(catalog_rows).to_csv(catalog_path, index=False)
    print("catalog saved:", catalog_path)

catalog_df = pd.DataFrame(catalog_rows)
catalog_df.to_csv(catalog_path, index=False)

success_df = catalog_df[catalog_df["status"] == "success"].copy()

print("\nFinished cases.")
display(catalog_df)

if len(success_df) == 0:
    raise RuntimeError("No successful cases. Stop here and inspect the error files.")


# ============================================================
# 9. Store successful results in memory
# ============================================================

# Since we just ran the models, keep result files saved and reload only if needed.
def load_result_npz(path):
    data = np.load(path, allow_pickle=True)
    result = {}

    for key in data.files:
        arr = data[key]

        if arr.shape == ():
            try:
                result[key] = arr.item()
            except Exception:
                result[key] = arr
        else:
            result[key] = arr

    return result


results = {}

for _, row in success_df.iterrows():
    label = row["label"]
    result = load_result_npz(row["result_npz"])
    results[label] = result

print("Loaded successful results:", list(results.keys()))


# ============================================================
# 10. Derived diagnostics
# ============================================================

def compute_radius_profiles(result):
    Pbar = get_result_array(result, "Pbar")
    n = get_result_array(result, "n")
    m_bin = get_result_array(result, "m_bin")
    r_nm = get_result_array(result, "r_compact_nm")

    if Pbar is None or n is None or m_bin is None or r_nm is None:
        raise RuntimeError("Missing one of Pbar, n, m_bin, r_compact_nm.")

    mass_density_bin = n * m_bin[None, :]
    number_density = np.sum(n, axis=1)
    mass_density = np.sum(mass_density_bin, axis=1)

    number_radius = np.sum(n * r_nm[None, :], axis=1) / np.maximum(
        number_density,
        1e-300,
    )

    mass_radius = np.sum(mass_density_bin * r_nm[None, :], axis=1) / np.maximum(
        mass_density,
        1e-300,
    )

    active_radius = np.full_like(Pbar, np.nan, dtype=float)

    for k in range(len(Pbar)):
        md = mass_density_bin[k, :]
        if np.nanmax(md) <= 0:
            continue

        active = np.where(md > np.nanmax(md) * 1e-8)[0]

        if active.size > 0:
            active_radius[k] = r_nm[active[-1]]

    return {
        "Pbar": Pbar,
        "n": n,
        "m_bin": m_bin,
        "r_nm": r_nm,
        "mass_density_bin": mass_density_bin,
        "number_density": number_density,
        "mass_density": mass_density,
        "number_radius_nm": number_radius,
        "mass_radius_nm": mass_radius,
        "active_radius_nm": active_radius,
    }


def compute_timescales(result):
    Pbar = get_result_array(result, "Pbar")
    H = get_result_array(result, "H")
    Kzz = get_result_array(result, "Kzz")
    dz = get_result_array(result, "dz")
    v_set = get_result_array(result, "v_set")
    n = get_result_array(result, "n")
    m_bin = get_result_array(result, "m_bin")
    r_nm = get_result_array(result, "r_compact_nm")
    K_coag = get_result_array(result, "K_coag")
    k_therm = get_result_array(result, "thermal_rate")

    if k_therm is None:
        k_therm = get_result_array(result, "k_therm")

    if H is None:
        H = np.ones_like(Pbar) * np.nanmedian(dz)

    if Kzz is None:
        Kzz = np.ones_like(Pbar) * np.nan

    tau_mix = H**2 / np.maximum(Kzz, 1e-300)

    tau_set_bins = {}
    tau_set_mass_weighted = None

    if v_set is not None and r_nm is not None:
        selected_r = np.array([1.0, 10.0, 30.0, 100.0, 1000.0])
        selected_bins = []

        for rr in selected_r:
            selected_bins.append(int(np.argmin(np.abs(r_nm - rr))))

        v_safe = np.maximum(v_set, 1e-300)

        for i in selected_bins:
            tau_set_bins[int(i)] = H / v_safe[:, i]

        if n is not None and m_bin is not None:
            mass_bin = n * m_bin[None, :]
            v_mass = np.sum(mass_bin * v_set, axis=1) / np.maximum(
                np.sum(mass_bin, axis=1),
                1e-300,
            )
            tau_set_mass_weighted = H / np.maximum(v_mass, 1e-300)
    else:
        selected_bins = []

    tau_coag = None

    if K_coag is not None and n is not None:
        try:
            freq_i = np.einsum("kij,kj->ki", K_coag, n)
            ntot = np.sum(n, axis=1)
            freq_eff = np.sum(n * freq_i, axis=1) / np.maximum(ntot, 1e-300)
            tau_coag = 1.0 / np.maximum(freq_eff, 1e-300)
        except Exception as exc:
            print("Could not compute coagulation timescale:", repr(exc))

    tau_therm = None

    if k_therm is not None:
        tau_therm = 1.0 / np.maximum(k_therm, 1e-300)

    return {
        "Pbar": Pbar,
        "tau_mix": tau_mix,
        "tau_set_bins": tau_set_bins,
        "tau_set_mass_weighted": tau_set_mass_weighted,
        "tau_coag": tau_coag,
        "tau_therm": tau_therm,
        "selected_bins": selected_bins,
    }


derived = {}
timescales = {}

for label, result in results.items():
    derived[label] = compute_radius_profiles(result)
    timescales[label] = compute_timescales(result)


# ============================================================
# 11. Figure folders
# ============================================================

fig_root = ensure_dir(OUTPUT_ROOT / "figures")

setup_dir = ensure_dir(fig_root / "setup")
history_dir = ensure_dir(fig_root / "history")
profile_dir = ensure_dir(fig_root / "profiles")
map_dir = ensure_dir(fig_root / "pressure_radius_maps")
radius_dir = ensure_dir(fig_root / "radius_pressure")
timescale_dir = ensure_dir(fig_root / "timescales")
settling_dir = ensure_dir(fig_root / "settling")
sizedist_dir = ensure_dir(fig_root / "size_distributions")
opacity_dir = ensure_dir(fig_root / "opacity")
summary_dir = ensure_dir(fig_root / "summary")


# ============================================================
# 12. Setup plots: P-T and Kzz
# ============================================================

first_label = list(results.keys())[0]
first_result = results[first_label]

Pbar = get_result_array(first_result, "Pbar")
T = get_result_array(first_result, "T")
Kzz = get_result_array(first_result, "Kzz")

fig, ax = plt.subplots()
ax.plot(T, Pbar)
ax.set_yscale("log")
ax.invert_yaxis()
ax.set_xlabel("Temperature [K]")
ax.set_ylabel("Pressure [bar]")
ax.set_title("Caligo background P-T profile")
savefig(fig, setup_dir / "background_PT_profile.png")
plt.show()

fig, ax = plt.subplots()
ax.plot(Kzz, Pbar)
ax.set_xscale("log")
ax.set_yscale("log")
ax.invert_yaxis()
ax.set_xlabel(r"$K_{zz}$ [cm$^2$ s$^{-1}$]")
ax.set_ylabel("Pressure [bar]")
ax.set_title(r"Caligo $K_{zz}$ profile")
savefig(fig, setup_dir / "background_Kzz_profile.png")
plt.show()


# ============================================================
# 13. History plots
# ============================================================

for _, row in success_df.iterrows():
    label = row["label"]
    hist = pd.read_csv(row["history_csv"])

    fig, ax = plt.subplots()
    t_days = hist["t_end"] / 86400.0

    ax.plot(t_days, hist["F_source"], marker="o", label="source")
    ax.plot(t_days, hist["F_sink"], marker="o", label="sink")
    ax.plot(t_days, hist["F_storage"], marker="o", label="storage")

    if "F_thermal" in hist.columns:
        ax.plot(t_days, hist["F_thermal"], marker=".", label="thermal")

    if "F_bottom" in hist.columns:
        ax.plot(t_days, hist["F_bottom"], marker=".", label="bottom")

    ax.set_xlabel("Time [days]")
    ax.set_ylabel(r"Flux [g cm$^{-2}$ s$^{-1}$]")
    ax.set_title(f"{label}: flux-budget convergence")
    ax.legend(fontsize=8)
    savefig(fig, history_dir / f"{label}_flux_budget.png")
    plt.show()

    fig, ax = plt.subplots()
    ax.plot(t_days, hist["column_mass"], marker="o")
    ax.set_xlabel("Time [days]")
    ax.set_ylabel(r"Haze column mass [g cm$^{-2}$]")
    ax.set_title(f"{label}: haze column mass")
    savefig(fig, history_dir / f"{label}_column_mass_history.png")
    plt.show()


# ============================================================
# 14. Source and final haze profiles
# ============================================================

fig, ax = plt.subplots()

for label, d in derived.items():
    ax.plot(
        np.maximum(d["mass_density"], 1e-300),
        d["Pbar"],
        label=f"{label}: haze",
    )

ax.set_xscale("log")
ax.set_yscale("log")
ax.invert_yaxis()
ax.set_xlabel(r"Final haze mass density [g cm$^{-3}$]")
ax.set_ylabel("Pressure [bar]")
ax.set_title("Final haze mass density profiles")
ax.legend(fontsize=8)
savefig(fig, profile_dir / "compare_final_haze_mass_density.png")
plt.show()


fig, ax = plt.subplots()

for label, result in results.items():
    Pbar = get_result_array(result, "Pbar")
    S_mass = get_result_array(result, "S_mass_density")

    if S_mass is None:
        continue

    ax.plot(np.maximum(S_mass, 1e-300), Pbar, label=label)

ax.set_xscale("log")
ax.set_yscale("log")
ax.invert_yaxis()
ax.set_xlabel(r"Haze source [g cm$^{-3}$ s$^{-1}$]")
ax.set_ylabel("Pressure [bar]")
ax.set_title("Haze source profiles")
ax.legend(fontsize=8)
savefig(fig, profile_dir / "compare_haze_source_profiles.png")
plt.show()


# ============================================================
# 15. Pressure-radius mass-density maps
# ============================================================

for label, d in derived.items():
    Pbar = d["Pbar"]
    r_nm = d["r_nm"]
    mass_density_bin = d["mass_density_bin"]

    Psort, msort = pressure_sort_top_to_bottom(Pbar, mass_density_bin)

    vmax = np.nanmax(msort)
    vmin = max(vmax * 1e-10, 1e-300)

    fig, ax = plt.subplots(figsize=(8, 5))

    im = ax.pcolormesh(
        r_nm,
        Psort,
        np.maximum(msort, vmin),
        shading="auto",
        norm=LogNorm(vmin=vmin, vmax=vmax),
    )

    ax.plot(
        d["mass_radius_nm"],
        Pbar,
        color="white",
        linewidth=2.0,
        label="mass-weighted radius",
    )

    ax.plot(
        d["number_radius_nm"],
        Pbar,
        color="black",
        linewidth=1.5,
        label="number-weighted radius",
    )

    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.invert_yaxis()
    ax.set_xlabel("Compact particle radius [nm]")
    ax.set_ylabel("Pressure [bar]")
    ax.set_title(f"{label}: haze mass density in pressure-radius space")
    ax.legend(fontsize=8)

    cb = plt.colorbar(im, ax=ax)
    cb.set_label(r"Haze mass density by bin [g cm$^{-3}$]")

    savefig(fig, map_dir / f"{label}_pressure_radius_mass_density.png")
    plt.show()


# ============================================================
# 16. Radius versus pressure plots
# ============================================================

fig, ax = plt.subplots()

for label, d in derived.items():
    ax.plot(d["mass_radius_nm"], d["Pbar"], label=label)

ax.set_xscale("log")
ax.set_yscale("log")
ax.invert_yaxis()
ax.set_xlabel("Mass-weighted particle radius [nm]")
ax.set_ylabel("Pressure [bar]")
ax.set_title("Mass-weighted particle radius versus pressure")
ax.legend(fontsize=8)
savefig(fig, radius_dir / "compare_mass_weighted_radius_pressure.png")
plt.show()


fig, ax = plt.subplots()

for label, d in derived.items():
    ax.plot(d["number_radius_nm"], d["Pbar"], label=label)

ax.set_xscale("log")
ax.set_yscale("log")
ax.invert_yaxis()
ax.set_xlabel("Number-weighted particle radius [nm]")
ax.set_ylabel("Pressure [bar]")
ax.set_title("Number-weighted particle radius versus pressure")
ax.legend(fontsize=8)
savefig(fig, radius_dir / "compare_number_weighted_radius_pressure.png")
plt.show()


fig, ax = plt.subplots()

for label, d in derived.items():
    ax.plot(d["mass_radius_nm"], d["Pbar"], label=f"{label}: mass")
    ax.plot(d["number_radius_nm"], d["Pbar"], linestyle="--", label=f"{label}: number")

ax.set_xscale("log")
ax.set_yscale("log")
ax.invert_yaxis()
ax.set_xlabel("Particle radius [nm]")
ax.set_ylabel("Pressure [bar]")
ax.set_title("Number- and mass-weighted radius versus pressure")
ax.legend(fontsize=7)
savefig(fig, radius_dir / "compare_number_and_mass_radius_pressure.png")
plt.show()


# ============================================================
# 17. Timescale plots
# ============================================================

for label, ts in timescales.items():
    Pbar = ts["Pbar"]

    fig, ax = plt.subplots(figsize=(7.5, 5.5))

    ax.plot(
        ts["tau_mix"],
        Pbar,
        label=r"$\tau_{\rm mix}=H^2/K_{zz}$",
        linewidth=2,
    )

    if ts["tau_set_mass_weighted"] is not None:
        ax.plot(
            ts["tau_set_mass_weighted"],
            Pbar,
            label=r"$\tau_{\rm set}$ mass-weighted",
            linewidth=2,
        )

    if ts["tau_coag"] is not None:
        ax.plot(
            ts["tau_coag"],
            Pbar,
            label=r"$\tau_{\rm coag}$ effective",
            linewidth=2,
        )

    if ts["tau_therm"] is not None:
        ax.plot(
            ts["tau_therm"],
            Pbar,
            label=r"$\tau_{\rm therm}$",
            linewidth=2,
        )

    for val, name in [
        (3600.0, "1 hr"),
        (86400.0, "1 d"),
        (86400.0 * 30.0, "30 d"),
        (86400.0 * 365.0, "1 yr"),
    ]:
        ax.axvline(val, linestyle=":", linewidth=0.8)
        ax.text(
            val,
            np.nanmax(Pbar),
            name,
            rotation=90,
            va="bottom",
            ha="right",
            fontsize=8,
        )

    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.invert_yaxis()
    ax.set_xlabel("Timescale [s]")
    ax.set_ylabel("Pressure [bar]")
    ax.set_title(f"{label}: microphysical and transport timescales")
    ax.legend(fontsize=8)
    savefig(fig, timescale_dir / f"{label}_timescales_pressure.png")
    plt.show()


fig, ax = plt.subplots(figsize=(7.5, 5.5))

for label, ts in timescales.items():
    if ts["tau_set_mass_weighted"] is not None:
        ax.plot(
            ts["tau_set_mass_weighted"],
            ts["Pbar"],
            label=f"{label}: settling",
        )

ax.set_xscale("log")
ax.set_yscale("log")
ax.invert_yaxis()
ax.set_xlabel("Mass-weighted settling timescale [s]")
ax.set_ylabel("Pressure [bar]")
ax.set_title("Comparison of mass-weighted settling timescale")
ax.legend(fontsize=8)
savefig(fig, timescale_dir / "compare_mass_weighted_settling_timescale.png")
plt.show()


# ============================================================
# 18. Settling speed by particle size
# ============================================================

for label, result in results.items():
    Pbar = get_result_array(result, "Pbar")
    v_set = get_result_array(result, "v_set")
    r_nm = get_result_array(result, "r_compact_nm")

    if v_set is None or r_nm is None:
        continue

    selected_r = [1.0, 10.0, 30.0, 100.0, 1000.0]

    fig, ax = plt.subplots()

    for rr in selected_r:
        i = int(np.argmin(np.abs(r_nm - rr)))
        ax.plot(
            np.maximum(v_set[:, i], 1e-300),
            Pbar,
            label=f"{r_nm[i]:.1g} nm",
        )

    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.invert_yaxis()
    ax.set_xlabel(r"Settling speed [cm s$^{-1}$]")
    ax.set_ylabel("Pressure [bar]")
    ax.set_title(f"{label}: settling speed by particle size")
    ax.legend(title="radius", fontsize=8)
    savefig(fig, settling_dir / f"{label}_settling_speed_selected_sizes.png")
    plt.show()


# ============================================================
# 19. Size distributions at selected pressures
# ============================================================

selected_pressures = np.array([1e-8, 1e-6, 1e-4, 1e-2, 1e0])

for label, d in derived.items():
    Pbar = d["Pbar"]
    r_nm = d["r_nm"]
    n = d["n"]
    mass_density_bin = d["mass_density_bin"]

    fig, ax = plt.subplots()

    for p0 in selected_pressures:
        k = int(np.argmin(np.abs(np.log10(Pbar) - np.log10(p0))))
        ax.loglog(
            r_nm,
            np.maximum(n[k, :], 1e-300),
            marker="o",
            label=f"{Pbar[k]:.1e} bar",
        )

    ax.set_xlabel("Compact particle radius [nm]")
    ax.set_ylabel(r"Number density by bin [cm$^{-3}$]")
    ax.set_title(f"{label}: number-size distributions")
    ax.legend(fontsize=8)
    savefig(fig, sizedist_dir / f"{label}_number_size_distributions.png")
    plt.show()

    fig, ax = plt.subplots()

    for p0 in selected_pressures:
        k = int(np.argmin(np.abs(np.log10(Pbar) - np.log10(p0))))
        ax.loglog(
            r_nm,
            np.maximum(mass_density_bin[k, :], 1e-300),
            marker="o",
            label=f"{Pbar[k]:.1e} bar",
        )

    ax.set_xlabel("Compact particle radius [nm]")
    ax.set_ylabel(r"Mass density by bin [g cm$^{-3}$]")
    ax.set_title(f"{label}: mass-size distributions")
    ax.legend(fontsize=8)
    savefig(fig, sizedist_dir / f"{label}_mass_size_distributions.png")
    plt.show()


# ============================================================
# 20. Direct-Mie opacity plots
# ============================================================

fig, ax = plt.subplots(figsize=(8, 5))

for _, row in success_df.iterrows():
    tau_csv = row.get("tau_csv", None)

    if not isinstance(tau_csv, str) or not Path(tau_csv).exists():
        continue

    tau_df = pd.read_csv(tau_csv)

    ax.loglog(
        tau_df["wave_um"],
        np.maximum(tau_df["slant_tau_ext"], 1e-300),
        label=row["label"],
    )

ax.axhline(1.0, linestyle="--", linewidth=1.0)
ax.set_xlabel("Wavelength [µm]")
ax.set_ylabel(r"Direct-Mie slant $\tau_{\rm ext}$")
ax.set_title("Direct-Mie haze opacity comparison")
ax.legend(fontsize=8)
savefig(fig, opacity_dir / "compare_direct_mie_slant_tau.png")
plt.show()


fig, ax = plt.subplots(figsize=(8, 5))

for _, row in success_df.iterrows():
    tau_csv = row.get("tau_csv", None)

    if not isinstance(tau_csv, str) or not Path(tau_csv).exists():
        continue

    tau_df = pd.read_csv(tau_csv)

    ax.semilogx(
        tau_df["wave_um"],
        tau_df["transmission_strength"],
        label=row["label"],
    )

ax.set_xlabel("Wavelength [µm]")
ax.set_ylabel(r"$1-e^{-\tau_{\rm slant}}$")
ax.set_title("Approximate haze transmission strength")
ax.legend(fontsize=8)
savefig(fig, opacity_dir / "compare_transmission_strength.png")
plt.show()


for _, row in success_df.iterrows():
    label = row["label"]
    layer_npz = row.get("layer_npz", None)

    if not isinstance(layer_npz, str) or not Path(layer_npz).exists():
        continue

    data = np.load(layer_npz)
    wave_um = data["wave_um"]
    Pbar = data["Pbar"]
    dtau_ext_layer = data["dtau_ext_layer"]

    Psort, dtausort = pressure_sort_top_to_bottom(Pbar, dtau_ext_layer)

    vmax = np.nanmax(dtausort)
    vmin = max(vmax * 1e-8, 1e-300)

    fig, ax = plt.subplots(figsize=(8, 5))

    im = ax.pcolormesh(
        wave_um,
        Psort,
        np.maximum(dtausort, vmin),
        shading="auto",
        norm=LogNorm(vmin=vmin, vmax=vmax),
    )

    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.invert_yaxis()
    ax.set_xlabel("Wavelength [µm]")
    ax.set_ylabel("Pressure [bar]")
    ax.set_title(f"{label}: layer contribution to vertical haze optical depth")

    cb = plt.colorbar(im, ax=ax)
    cb.set_label(r"$d\tau_{\rm ext}$ per layer")

    savefig(fig, opacity_dir / f"{label}_wavelength_pressure_dtau_map.png")
    plt.show()


fig, ax = plt.subplots(figsize=(8, 5))

for _, row in success_df.iterrows():
    tau_csv = row.get("tau_csv", None)

    if not isinstance(tau_csv, str) or not Path(tau_csv).exists():
        continue

    tau_df = pd.read_csv(tau_csv)

    ax.semilogx(
        tau_df["wave_um"],
        tau_df["p_eff_bar"],
        label=row["label"],
    )

ax.set_yscale("log")
ax.invert_yaxis()
ax.set_xlabel("Wavelength [µm]")
ax.set_ylabel("Extinction-weighted pressure [bar]")
ax.set_title("Where the haze opacity lives")
ax.legend(fontsize=8)
savefig(fig, opacity_dir / "compare_opacity_weighted_pressure.png")
plt.show()


# ============================================================
# 21. Source-flux summary plots
# ============================================================

sf = success_df.sort_values("target_haze_flux").copy()

fig, ax = plt.subplots()
ax.plot(sf["target_haze_flux"], sf["column_mass_g_cm2"], marker="o")
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel(r"Target haze source flux [g cm$^{-2}$ s$^{-1}$]")
ax.set_ylabel(r"Final haze column mass [g cm$^{-2}$]")
ax.set_title("Haze column mass versus source flux")
savefig(fig, summary_dir / "column_mass_vs_source_flux.png")
plt.show()


fig, ax = plt.subplots()
ax.plot(sf["target_haze_flux"], sf["mass_weighted_radius_nm"], marker="o")
ax.set_xscale("log")
ax.set_xlabel(r"Target haze source flux [g cm$^{-2}$ s$^{-1}$]")
ax.set_ylabel("Mass-weighted radius [nm]")
ax.set_title("Particle size versus source flux")
savefig(fig, summary_dir / "mass_weighted_radius_vs_source_flux.png")
plt.show()


if "max_slant_tau_ext" in sf.columns:
    fig, ax = plt.subplots()
    ax.plot(sf["target_haze_flux"], sf["max_slant_tau_ext"], marker="o")
    ax.axhline(1.0, linestyle="--", linewidth=1.0)
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.set_xlabel(r"Target haze source flux [g cm$^{-2}$ s$^{-1}$]")
    ax.set_ylabel(r"Max direct-Mie slant $\tau_{\rm ext}$")
    ax.set_title("Haze opacity versus source flux")
    savefig(fig, summary_dir / "max_slant_tau_vs_source_flux.png")
    plt.show()


if "max_transmission_strength" in sf.columns:
    fig, ax = plt.subplots()
    ax.plot(sf["target_haze_flux"], sf["max_transmission_strength"], marker="o")
    ax.set_xscale("log")
    ax.set_xlabel(r"Target haze source flux [g cm$^{-2}$ s$^{-1}$]")
    ax.set_ylabel(r"Max $1-e^{-\tau_{\rm slant}}$")
    ax.set_title("Transmission strength versus source flux")
    savefig(fig, summary_dir / "transmission_strength_vs_source_flux.png")
    plt.show()


# ============================================================
# 22. Save manifest
# ============================================================

manifest = {
    "project_root": str(PROJECT_ROOT),
    "example_script": str(EXAMPLE_SCRIPT),
    "output_root": str(OUTPUT_ROOT),
    "cases": CASES,
    "solver_settings": SOLVER_SETTINGS,
    "catalog_path": str(catalog_path),
    "successful_labels": list(success_df["label"]),
    "figure_directories": {
        "setup": str(setup_dir),
        "history": str(history_dir),
        "profiles": str(profile_dir),
        "pressure_radius_maps": str(map_dir),
        "radius_pressure": str(radius_dir),
        "timescales": str(timescale_dir),
        "settling": str(settling_dir),
        "size_distributions": str(sizedist_dir),
        "opacity": str(opacity_dir),
        "summary": str(summary_dir),
    },
}

manifest_path = OUTPUT_ROOT / "manifest.json"
write_json(manifest_path, manifest)

print("\nDONE.")
print("All outputs saved in:")
print(OUTPUT_ROOT)
print("Catalog:")
print(catalog_path)
print("Manifest:")
print(manifest_path)
```

```python
# ============================================================
# FIXED SAVED-RUN TIMESCALE + SPECTRUM DIAGNOSTICS
#
# This does NOT rerun the time-dependent haze solver.
#
# It:
#   1. Loads the saved final haze distributions from result_npz.
#   2. Rebuilds the microphysics context for each saved case.
#   3. Uses context arrays v_set, K_coag, thermal_rate, H, Kzz.
#   4. Computes real mixing, settling, coagulation, and thermal timescales.
#   5. Saves paper-style timescale plots.
#   6. Recreates the saved direct-Mie spectrum comparison.
# ============================================================

from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import traceback

# ------------------------------------------------------------
# 0. Paths
# ------------------------------------------------------------

PROJECT_ROOT = Path("/Users/svc553/Documents/Caligo").resolve()
OUTPUT_ROOT = PROJECT_ROOT / "output" / "formal_figures_flux_comparison_v2"
CATALOG_PATH = OUTPUT_ROOT / "run_catalog.csv"

DIAG_ROOT = OUTPUT_ROOT / "paper_timescale_spectrum_comparison_fixed"
TIMESCALE_DIR = DIAG_ROOT / "timescales"
SPECTRUM_DIR = DIAG_ROOT / "spectra"
TABLE_DIR = DIAG_ROOT / "tables"

for d in [DIAG_ROOT, TIMESCALE_DIR, SPECTRUM_DIR, TABLE_DIR]:
    d.mkdir(parents=True, exist_ok=True)

print("Using OUTPUT_ROOT:")
print(OUTPUT_ROOT)
print("\nUsing catalog:")
print(CATALOG_PATH)

if not CATALOG_PATH.exists():
    raise FileNotFoundError(f"Could not find catalog: {CATALOG_PATH}")


# ------------------------------------------------------------
# 1. Plot/save helpers
# ------------------------------------------------------------

plt.rcParams.update({
    "figure.figsize": (7.8, 5.3),
    "axes.grid": True,
    "grid.alpha": 0.28,
    "font.size": 11,
})


def savefig(fig, path):
    path = Path(path)
    path.parent.mkdir(parents=True, exist_ok=True)
    fig.savefig(path, dpi=220, bbox_inches="tight")
    print("saved:", path)


def load_npz_as_dict(path):
    data = np.load(path, allow_pickle=True)
    out = {}

    for key in data.files:
        arr = data[key]
        if arr.shape == ():
            try:
                out[key] = arr.item()
            except Exception:
                out[key] = arr
        else:
            out[key] = arr

    return out


def get_from_dicts(key, *dicts, default=None):
    for d in dicts:
        if isinstance(d, dict) and key in d:
            return d[key]
    return default


def get_first_key(dicts, keys, default=None):
    for key in keys:
        for d in dicts:
            if isinstance(d, dict) and key in d:
                return d[key], key
    return default, None


def clean_positive(x, floor=1e-300):
    x = np.asarray(x, dtype=float)
    y = np.where(np.isfinite(x) & (x > floor), x, np.nan)
    return y


def pressure_plot_setup(ax, xlabel, title):
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.invert_yaxis()
    ax.set_xlabel(xlabel)
    ax.set_ylabel("Pressure [bar]")
    ax.set_title(title)


def local_scale_height_from_pz(Pbar, z):
    """
    Estimate H = -1 / d ln(P) / dz from saved/rebuilt grids.
    This is a fallback only if ctx['H'] is missing.
    """
    Pbar = np.asarray(Pbar, dtype=float)
    z = np.asarray(z, dtype=float)

    order = np.argsort(z)
    z_sort = z[order]
    P_sort = Pbar[order]

    lnP = np.log(np.maximum(P_sort, 1e-300))
    dlnP_dz = np.gradient(lnP, z_sort)
    H_sort = 1.0 / np.maximum(np.abs(dlnP_dz), 1e-300)

    H = np.empty_like(H_sort)
    H[order] = H_sort
    return H


def maybe_display(df):
    try:
        display(df)
    except Exception:
        print(df)


# ------------------------------------------------------------
# 2. Load saved run catalog and saved final results
# ------------------------------------------------------------

catalog = pd.read_csv(CATALOG_PATH)
success = catalog[catalog["status"].astype(str).str.lower() == "success"].copy()

if len(success) == 0:
    raise RuntimeError("No successful cases found in catalog.")

print("\nSuccessful cases:")
cols = [
    "label",
    "converged",
    "target_haze_flux",
    "column_mass_g_cm2",
    "mass_weighted_radius_nm",
    "max_slant_tau_ext",
    "max_transmission_strength",
]
maybe_display(success[[c for c in cols if c in success.columns]])

results = {}

for _, row in success.iterrows():
    label = row["label"]
    result_npz = Path(row["result_npz"])

    if not result_npz.exists():
        raise FileNotFoundError(f"Missing saved result for {label}: {result_npz}")

    results[label] = load_npz_as_dict(result_npz)

print("\nLoaded saved result labels:")
print(list(results.keys()))


# ------------------------------------------------------------
# 3. Rebuild context for each saved case
# ------------------------------------------------------------

if "base_case" not in globals():
    raise RuntimeError(
        "base_case is not defined. Run the setup/photochem part of the notebook first. "
        "This block intentionally avoids rerunning the expensive solver, but it needs "
        "the existing Caligo case object to rebuild the microphysics context."
    )

if "attach_science_settings" not in globals():
    raise RuntimeError(
        "attach_science_settings is not defined. Run the full notebook setup cell first."
    )

if "reset_caligo_runtime_cache" not in globals():
    raise RuntimeError(
        "reset_caligo_runtime_cache is not defined. Run the full notebook setup cell first."
    )

if "ensure_source_info_built" not in globals():
    raise RuntimeError(
        "ensure_source_info_built is not defined. Run the full notebook setup cell first."
    )


def rebuild_context_for_saved_case(case_obj, row):
    """
    Rebuilds microphysics context for a saved run without running the
    time-dependent solver.
    """
    label = row["label"]

    print("\n" + "=" * 72)
    print("Rebuilding microphysics context for:", label)
    print("=" * 72)

    reset_caligo_runtime_cache(case_obj)

    attach_science_settings(
        case_obj,
        target_haze_flux=float(row["target_haze_flux"]),
        gauss_peak_bar=float(row["gauss_peak_bar"]),
        gauss_sigma_dex=float(row["gauss_sigma_dex"]),
    )

    source_info = ensure_source_info_built(case_obj)

    # This should be fast. It rebuilds context arrays, not the full time-dependent run.
    ctx = case_obj.prepare_microphysics(
        climate_out=None,
        with_coagulation=True,
    )

    # Some local versions may return a wrapper.
    if isinstance(ctx, dict) and "ctx" in ctx and isinstance(ctx["ctx"], dict):
        ctx = ctx["ctx"]

    if not isinstance(ctx, dict):
        raise RuntimeError(f"{label}: prepare_microphysics did not return a dict context.")

    print("Context rebuilt.")
    print("Available key examples:", list(ctx.keys())[:25])

    return ctx


contexts = {}

for _, row in success.iterrows():
    label = row["label"]
    try:
        contexts[label] = rebuild_context_for_saved_case(base_case, row)
    except Exception as exc:
        print(f"FAILED context rebuild for {label}: {repr(exc)}")
        print(traceback.format_exc())


if len(contexts) == 0:
    raise RuntimeError("No contexts could be rebuilt.")


# ------------------------------------------------------------
# 4. Compute real timescales from saved final n + rebuilt context
# ------------------------------------------------------------

def compute_timescales_from_saved_result_and_context(label, result, ctx):
    """
    Uses:
      saved final n from result
      context v_set, K_coag, thermal_rate, H, Kzz from rebuilt context

    Returns a dataframe with one row per pressure level.
    """

    Pbar = np.asarray(get_from_dicts("Pbar", result, ctx), dtype=float)
    n = np.asarray(get_from_dicts("n", result), dtype=float)

    m_bin = np.asarray(get_from_dicts("m_bin", result, ctx), dtype=float)
    r_nm = np.asarray(get_from_dicts("r_compact_nm", result, ctx), dtype=float)

    Kzz = get_from_dicts("Kzz", ctx, result)
    H = get_from_dicts("H", ctx, result)
    z = get_from_dicts("z", ctx, result)
    dz = get_from_dicts("dz", ctx, result)

    v_set, v_set_key = get_first_key(
        [ctx, result],
        ["v_set", "v_settle", "settling_velocity", "vfall", "v_fall"],
        default=None,
    )

    K_coag, K_coag_key = get_first_key(
        [ctx, result],
        ["K_coag", "coag_kernel", "Kij", "coagulation_kernel"],
        default=None,
    )

    thermal_rate, thermal_key = get_first_key(
        [ctx, result],
        [
            "thermal_rate",
            "k_therm",
            "k_thermal",
            "therm_destroy_rate",
            "thermal_destruction_rate",
            "k_loss_thermal",
        ],
        default=None,
    )

    print(f"\n{label}: timescale source arrays")
    print("  H exists:", H is not None)
    print("  Kzz exists:", Kzz is not None)
    print("  v_set key:", v_set_key)
    print("  K_coag key:", K_coag_key)
    print("  thermal key:", thermal_key)

    if H is None:
        if z is not None:
            H = local_scale_height_from_pz(Pbar, z)
            print("  H fallback: computed from P-z gradient.")
        elif dz is not None:
            H = np.ones_like(Pbar) * np.nanmedian(np.asarray(dz, dtype=float))
            print("  H fallback: median dz.")
        else:
            raise RuntimeError(f"{label}: no H, z, or dz available.")

    H = np.asarray(H, dtype=float)

    if Kzz is None:
        tau_mix = np.full_like(Pbar, np.nan, dtype=float)
    else:
        Kzz = np.asarray(Kzz, dtype=float)
        tau_mix = H**2 / np.maximum(Kzz, 1e-300)

    mass_bin = n * m_bin[None, :]
    mass_density = np.sum(mass_bin, axis=1)
    number_density = np.sum(n, axis=1)

    # Only hide truly empty layers; do not hide real low-density regions too aggressively.
    active_mask = mass_density > max(np.nanmax(mass_density) * 1e-30, 1e-300)

    # -------------------------
    # Settling timescale
    # -------------------------
    tau_set_mass = np.full_like(Pbar, np.nan, dtype=float)

    selected_settling = {}

    if v_set is not None:
        v_set = np.asarray(v_set, dtype=float)

        if v_set.shape == n.shape:
            v_mass = np.sum(mass_bin * v_set, axis=1) / np.maximum(mass_density, 1e-300)
            tau_set_mass = H / np.maximum(v_mass, 1e-300)
            tau_set_mass[~active_mask] = np.nan

            selected_radii_nm = [1.0, 3.0, 10.0, 30.0, 100.0, 300.0, 1000.0]

            for rr in selected_radii_nm:
                i = int(np.argmin(np.abs(r_nm - rr)))
                tau_i = H / np.maximum(v_set[:, i], 1e-300)
                tau_i[~active_mask] = np.nan
                selected_settling[f"tau_set_r{r_nm[i]:.3g}nm_s"] = tau_i
        else:
            print("  WARNING: v_set shape does not match n shape:", v_set.shape, n.shape)

    # -------------------------
    # Coagulation timescale
    # -------------------------
    tau_coag = np.full_like(Pbar, np.nan, dtype=float)

    if K_coag is not None:
        K_coag = np.asarray(K_coag, dtype=float)

        if K_coag.ndim == 3 and K_coag.shape[0] == n.shape[0]:
            # collision frequency for particles in bin i:
            # f_i = sum_j K_ij n_j
            freq_i = np.einsum("kij,kj->ki", K_coag, n)

            # number-weighted effective frequency:
            # <f> = sum_i n_i f_i / sum_i n_i
            freq_eff = np.sum(n * freq_i, axis=1) / np.maximum(number_density, 1e-300)

            tau_coag = 1.0 / np.maximum(freq_eff, 1e-300)
            tau_coag[~active_mask] = np.nan
        else:
            print("  WARNING: K_coag shape unexpected:", K_coag.shape)

    # -------------------------
    # Thermal destruction timescale
    # -------------------------
    tau_therm = np.full_like(Pbar, np.nan, dtype=float)

    if thermal_rate is not None:
        thermal_rate = np.asarray(thermal_rate, dtype=float)

        if thermal_rate.ndim == 1 and len(thermal_rate) == len(Pbar):
            tau_therm = 1.0 / np.maximum(thermal_rate, 1e-300)
            tau_therm[~active_mask] = np.nan
        else:
            print("  WARNING: thermal_rate shape unexpected:", thermal_rate.shape)

    df = pd.DataFrame({
        "label": label,
        "Pbar": Pbar,
        "H_cm": H,
        "Kzz_cm2_s": Kzz if Kzz is not None else np.nan,
        "tau_mix_s": tau_mix,
        "tau_set_mass_s": tau_set_mass,
        "tau_coag_s": tau_coag,
        "tau_therm_s": tau_therm,
        "mass_density_g_cm3": mass_density,
        "number_density_cm3": number_density,
    })

    for key, val in selected_settling.items():
        df[key] = val

    return df


timescale_tables = []

for label, result in results.items():
    if label not in contexts:
        continue

    df = compute_timescales_from_saved_result_and_context(
        label=label,
        result=result,
        ctx=contexts[label],
    )

    timescale_tables.append(df)

timescale_all = pd.concat(timescale_tables, ignore_index=True)

timescale_csv = TABLE_DIR / "timescale_comparison_fixed_long.csv"
timescale_all.to_csv(timescale_csv, index=False)

print("\nSaved fixed timescale table:")
print(timescale_csv)

print("\nNaN check by column:")
print(timescale_all[["tau_mix_s", "tau_set_mass_s", "tau_coag_s", "tau_therm_s"]].isna().mean())

try:
    display(timescale_all.head())
except Exception:
    print(timescale_all.head())


# ------------------------------------------------------------
# 5. Timescale plots
# ------------------------------------------------------------

# 5a. One full timescale plot per case.
for label in timescale_all["label"].unique():
    df = timescale_all[timescale_all["label"] == label].copy()

    fig, ax = plt.subplots(figsize=(7.8, 5.5))

    curves = [
        ("tau_mix_s", r"mixing: $H^2/K_{zz}$"),
        ("tau_set_mass_s", r"settling: $H/v_{\rm set}$, mass-weighted"),
        ("tau_coag_s", r"coagulation: $1/(K n)$"),
        ("tau_therm_s", r"thermal destruction"),
    ]

    for col, lab in curves:
        if col not in df.columns:
            continue

        y = clean_positive(df[col])

        if np.isfinite(y).sum() == 0:
            print(f"{label}: skipping {col}, all NaN/nonpositive.")
            continue

        ax.plot(y, df["Pbar"], linewidth=2, label=lab)

    for t_ref, name in [
        (3600.0, "1 hr"),
        (86400.0, "1 d"),
        (86400.0 * 30.0, "30 d"),
        (86400.0 * 365.0, "1 yr"),
    ]:
        ax.axvline(t_ref, linestyle=":", linewidth=0.8)
        ax.text(
            t_ref,
            np.nanmax(df["Pbar"]),
            name,
            rotation=90,
            fontsize=8,
            va="bottom",
            ha="right",
        )

    pressure_plot_setup(
        ax,
        "Timescale [s]",
        f"{label}: microphysical and transport timescales",
    )

    ax.legend(fontsize=8)
    savefig(fig, TIMESCALE_DIR / f"{label}_timescales_fixed.png")
    plt.show()


# 5b. Comparison panels across cases.
comparison_specs = [
    ("tau_mix_s", r"Mixing timescale $H^2/K_{zz}$ [s]", "compare_mixing_timescale_fixed.png"),
    ("tau_set_mass_s", r"Mass-weighted settling timescale $H/v_{\rm set}$ [s]", "compare_mass_weighted_settling_timescale_fixed.png"),
    ("tau_coag_s", r"Effective coagulation timescale $1/(K n)$ [s]", "compare_coagulation_timescale_fixed.png"),
    ("tau_therm_s", r"Thermal destruction timescale [s]", "compare_thermal_timescale_fixed.png"),
]

for col, xlabel, fname in comparison_specs:
    fig, ax = plt.subplots(figsize=(7.8, 5.5))

    any_curve = False

    for label in timescale_all["label"].unique():
        df = timescale_all[timescale_all["label"] == label].copy()

        y = clean_positive(df[col])

        if np.isfinite(y).sum() == 0:
            continue

        ax.plot(y, df["Pbar"], linewidth=2, label=label)
        any_curve = True

    if not any_curve:
        plt.close(fig)
        print(f"Skipping {fname}; no finite curves.")
        continue

    pressure_plot_setup(ax, xlabel, col.replace("_", " "))
    ax.legend(fontsize=8)

    savefig(fig, TIMESCALE_DIR / fname)
    plt.show()


# 5c. Settling timescale by selected particle size, one plot per case.
for label in timescale_all["label"].unique():
    df = timescale_all[timescale_all["label"] == label].copy()

    selected_cols = [
        c for c in df.columns
        if c.startswith("tau_set_r") and c.endswith("nm_s")
    ]

    if len(selected_cols) == 0:
        continue

    fig, ax = plt.subplots(figsize=(7.8, 5.5))

    for col in selected_cols:
        y = clean_positive(df[col])

        if np.isfinite(y).sum() == 0:
            continue

        radius_label = col.replace("tau_set_r", "").replace("nm_s", " nm")

        ax.plot(y, df["Pbar"], linewidth=1.8, label=radius_label)

    pressure_plot_setup(
        ax,
        r"Settling timescale $H/v_{\rm set}$ [s]",
        f"{label}: settling timescale by particle size",
    )

    ax.legend(fontsize=8, title="radius")
    savefig(fig, TIMESCALE_DIR / f"{label}_settling_by_particle_size_fixed.png")
    plt.show()


# ------------------------------------------------------------
# 6. Spectrum comparison from saved direct-Mie products
# ------------------------------------------------------------

def find_tau_csv(row):
    tau_csv = row.get("tau_csv", "")

    if isinstance(tau_csv, str) and len(tau_csv) > 0 and Path(tau_csv).exists():
        return Path(tau_csv)

    run_dir = Path(row["run_dir"])
    candidates = sorted(run_dir.rglob("*direct_mie_tau.csv"))

    if len(candidates) > 0:
        return candidates[0]

    return None


spectra = []

for _, row in success.iterrows():
    label = row["label"]
    tau_csv = find_tau_csv(row)

    if tau_csv is None:
        print(f"{label}: no direct-Mie tau csv found.")
        continue

    tau = pd.read_csv(tau_csv)
    tau["label"] = label
    tau["tau_csv"] = str(tau_csv)

    if "transmission_strength" not in tau.columns and "slant_tau_ext" in tau.columns:
        tau["transmission_strength"] = 1.0 - np.exp(-tau["slant_tau_ext"])

    if "relative_slant_proxy" not in tau.columns and "slant_tau_ext" in tau.columns:
        tau_1um = np.interp(1.0, tau["wave_um"], tau["slant_tau_ext"])
        tau["relative_slant_proxy"] = np.log1p(tau["slant_tau_ext"]) - np.log1p(tau_1um)

    spectra.append(tau)

if len(spectra) == 0:
    raise RuntimeError("No direct-Mie spectrum files found.")

spectrum_all = pd.concat(spectra, ignore_index=True)

spectrum_csv = TABLE_DIR / "spectrum_comparison_fixed_long.csv"
spectrum_all.to_csv(spectrum_csv, index=False)

print("\nSaved spectrum table:")
print(spectrum_csv)


# 6a. Slant tau comparison
fig, ax = plt.subplots(figsize=(8.0, 5.2))

for label in spectrum_all["label"].unique():
    df = spectrum_all[spectrum_all["label"] == label].copy()

    ax.loglog(
        df["wave_um"],
        clean_positive(df["slant_tau_ext"]),
        linewidth=2,
        label=label,
    )

ax.axhline(1.0, linestyle="--", linewidth=1.0, label=r"$\tau=1$")
ax.set_xlabel("Wavelength [µm]")
ax.set_ylabel(r"Direct-Mie slant $\tau_{\rm ext}$")
ax.set_title("Direct-Mie haze opacity comparison")
ax.legend(fontsize=8)

savefig(fig, SPECTRUM_DIR / "compare_direct_mie_slant_tau_ext_fixed.png")
plt.show()


# 6b. Approximate transmission strength
fig, ax = plt.subplots(figsize=(8.0, 5.2))

for label in spectrum_all["label"].unique():
    df = spectrum_all[spectrum_all["label"] == label].copy()

    ax.semilogx(
        df["wave_um"],
        df["transmission_strength"],
        linewidth=2,
        label=label,
    )

ax.set_xlabel("Wavelength [µm]")
ax.set_ylabel(r"$1-e^{-\tau_{\rm slant}}$")
ax.set_title("Approximate haze transmission strength")
ax.legend(fontsize=8)

savefig(fig, SPECTRUM_DIR / "compare_haze_transmission_strength_fixed.png")
plt.show()


# 6c. Absorption/scattering/extinction for each case
for label in spectrum_all["label"].unique():
    df = spectrum_all[spectrum_all["label"] == label].copy()

    fig, ax = plt.subplots(figsize=(8.0, 5.2))

    for col, lab in [
        ("slant_tau_ext", r"$\tau_{\rm ext}$"),
        ("slant_tau_abs", r"$\tau_{\rm abs}$"),
        ("slant_tau_sca", r"$\tau_{\rm sca}$"),
    ]:
        if col not in df.columns:
            continue

        ax.loglog(
            df["wave_um"],
            clean_positive(df[col]),
            linewidth=2,
            label=lab,
        )

    ax.axhline(1.0, linestyle="--", linewidth=1.0)
    ax.set_xlabel("Wavelength [µm]")
    ax.set_ylabel("Slant optical depth")
    ax.set_title(f"{label}: direct-Mie absorption/scattering/extinction")
    ax.legend(fontsize=8)

    savefig(fig, SPECTRUM_DIR / f"{label}_slant_abs_sca_ext_fixed.png")
    plt.show()


# 6d. Relative spectral-shape proxy
fig, ax = plt.subplots(figsize=(8.0, 5.2))

for label in spectrum_all["label"].unique():
    df = spectrum_all[spectrum_all["label"] == label].copy()

    ax.semilogx(
        df["wave_um"],
        df["relative_slant_proxy"],
        linewidth=2,
        label=label,
    )

ax.axhline(0.0, linewidth=1.0)
ax.set_xlabel("Wavelength [µm]")
ax.set_ylabel(r"$\log(1+\tau_{\rm slant})-\log(1+\tau_{\rm slant,1\mu m})$")
ax.set_title("Relative haze spectral-shape proxy")
ax.legend(fontsize=8)

savefig(fig, SPECTRUM_DIR / "compare_relative_slant_spectral_shape_fixed.png")
plt.show()


# 6e. Extinction-weighted pressure
if "p_eff_bar" in spectrum_all.columns:
    fig, ax = plt.subplots(figsize=(8.0, 5.2))

    for label in spectrum_all["label"].unique():
        df = spectrum_all[spectrum_all["label"] == label].copy()

        ax.semilogx(
            df["wave_um"],
            df["p_eff_bar"],
            linewidth=2,
            label=label,
        )

    ax.set_yscale("log")
    ax.invert_yaxis()
    ax.set_xlabel("Wavelength [µm]")
    ax.set_ylabel("Extinction-weighted pressure [bar]")
    ax.set_title("Where the haze opacity lives")
    ax.legend(fontsize=8)

    savefig(fig, SPECTRUM_DIR / "compare_extinction_weighted_pressure_fixed.png")
    plt.show()


# ------------------------------------------------------------
# 7. Summary table
# ------------------------------------------------------------

summary_rows = []

for label in timescale_all["label"].unique():
    tdf = timescale_all[timescale_all["label"] == label].copy()
    sdf = spectrum_all[spectrum_all["label"] == label].copy()

    # representative location: peak haze mass density
    k_peak = int(np.nanargmax(tdf["mass_density_g_cm3"].values))

    row_cat = success[success["label"] == label].iloc[0]

    summary_rows.append({
        "label": label,
        "target_haze_flux": float(row_cat["target_haze_flux"]),
        "column_mass_g_cm2": float(row_cat["column_mass_g_cm2"]),
        "mass_weighted_radius_nm": float(row_cat["mass_weighted_radius_nm"]),
        "max_slant_tau_ext": float(row_cat["max_slant_tau_ext"]),
        "max_transmission_strength": float(row_cat["max_transmission_strength"]),
        "p_mass_peak_bar": float(tdf["Pbar"].values[k_peak]),
        "tau_mix_at_mass_peak_s": float(tdf["tau_mix_s"].values[k_peak]),
        "tau_set_at_mass_peak_s": float(tdf["tau_set_mass_s"].values[k_peak]),
        "tau_coag_at_mass_peak_s": float(tdf["tau_coag_s"].values[k_peak]),
        "tau_therm_at_mass_peak_s": float(tdf["tau_therm_s"].values[k_peak]),
    })

summary_df = pd.DataFrame(summary_rows)

summary_csv = TABLE_DIR / "fixed_timescale_spectrum_summary.csv"
summary_df.to_csv(summary_csv, index=False)

print("\nFixed diagnostic summary:")
try:
    display(summary_df)
except Exception:
    print(summary_df)

print("\nSaved summary:")
print(summary_csv)

print("\nDONE.")
print("Fixed diagnostic outputs saved in:")
print(DIAG_ROOT)
```

```python
# ============================================================
# PAPER FIGURE PRODUCTION RUN:
# Source flux -> haze mass/size -> timescales -> opacity spectra
#
# This block:
#   1. Searches old saved catalogs for already-computed runs.
#   2. Reuses existing successful runs when possible.
#   3. Runs only missing flux-ladder cases until a walltime budget is reached.
#   4. Rebuilds microphysics contexts for real timescale diagnostics.
#   5. Makes paper-style figures that connect:
#        source flux
#        final haze mass
#        particle sizes
#        mixing/settling/coagulation/thermal timescales
#        direct-Mie spectral opacity
#
# It does NOT change your science settings except source flux.
# ============================================================

from pathlib import Path
import os
import time
import json
import traceback
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm


# ============================================================
# 0. Required notebook-state checks
# ============================================================

required_globals = [
    "base_case",
    "run_one_case",
    "attach_science_settings",
    "reset_caligo_runtime_cache",
    "ensure_source_info_built",
]

missing_globals = [g for g in required_globals if g not in globals()]

if missing_globals:
    raise RuntimeError(
        "Missing required notebook objects/functions: "
        + ", ".join(missing_globals)
        + "\nRun the full Caligo setup cell first."
    )


# ============================================================
# 1. Paths and runtime budget
# ============================================================

PROJECT_ROOT = Path("/Users/svc553/Documents/Caligo").resolve()

LADDER_ROOT = PROJECT_ROOT / "output" / "paper_flux_ladder_source_to_spectrum"
LADDER_ROOT.mkdir(parents=True, exist_ok=True)

FIG_ROOT = LADDER_ROOT / "figures"
TABLE_ROOT = LADDER_ROOT / "tables"
RUN_ROOT = LADDER_ROOT / "runs"

for d in [FIG_ROOT, TABLE_ROOT, RUN_ROOT]:
    d.mkdir(parents=True, exist_ok=True)

# You said you have about 15 hours. Keep a buffer so it does not start
# a long final case too close to the limit.
WALLTIME_HOURS = 14.5
WALLTIME_SECONDS = WALLTIME_HOURS * 3600.0

RUN_NEW_MISSING_CASES = True
REUSE_EXISTING_SUCCESSFUL_CASES = True

# These are ordered by scientific usefulness for a flux-ladder figure.
# Existing 1e-13 and 2e-13 cases should be reused if found.
FLUX_LADDER_CASES = [
    {
        "label": "flux5e-14_peak1e-7",
        "target_haze_flux": 5.0e-14,
        "gauss_peak_bar": 1.0e-7,
        "gauss_sigma_dex": 0.50,
        "estimated_runtime_min": 90.0,
    },
    {
        "label": "flux1e-13_peak1e-7",
        "target_haze_flux": 1.0e-13,
        "gauss_peak_bar": 1.0e-7,
        "gauss_sigma_dex": 0.50,
        "estimated_runtime_min": 145.0,
    },
    {
        "label": "flux1p5e-13_peak1e-7",
        "target_haze_flux": 1.5e-13,
        "gauss_peak_bar": 1.0e-7,
        "gauss_sigma_dex": 0.50,
        "estimated_runtime_min": 230.0,
    },
    {
        "label": "flux2e-13_peak1e-7",
        "target_haze_flux": 2.0e-13,
        "gauss_peak_bar": 1.0e-7,
        "gauss_sigma_dex": 0.50,
        "estimated_runtime_min": 335.0,
    },
    {
        "label": "flux2p5e-13_peak1e-7",
        "target_haze_flux": 2.5e-13,
        "gauss_peak_bar": 1.0e-7,
        "gauss_sigma_dex": 0.50,
        "estimated_runtime_min": 450.0,
    },
    # Only runs if there is time. If you already have it saved elsewhere,
    # the code will reuse it instead.
    {
        "label": "flux3e-13_peak1e-7",
        "target_haze_flux": 3.0e-13,
        "gauss_peak_bar": 1.0e-7,
        "gauss_sigma_dex": 0.50,
        "estimated_runtime_min": 565.0,
    },
]

MASTER_CATALOG_PATH = TABLE_ROOT / "flux_ladder_master_catalog.csv"

print("LADDER_ROOT:", LADDER_ROOT)
print("WALLTIME_HOURS:", WALLTIME_HOURS)


# ============================================================
# 2. Plot helpers
# ============================================================

plt.rcParams.update({
    "figure.figsize": (7.8, 5.2),
    "axes.grid": True,
    "grid.alpha": 0.28,
    "font.size": 11,
})


def savefig(fig, path):
    path = Path(path)
    path.parent.mkdir(parents=True, exist_ok=True)
    fig.savefig(path, dpi=240, bbox_inches="tight")
    print("saved:", path)


def safe_float(x, default=np.nan):
    try:
        return float(x)
    except Exception:
        return default


def load_npz_as_dict(path):
    data = np.load(path, allow_pickle=True)
    out = {}

    for key in data.files:
        arr = data[key]
        if arr.shape == ():
            try:
                out[key] = arr.item()
            except Exception:
                out[key] = arr
        else:
            out[key] = arr

    return out


def clean_positive(x, floor=1e-300):
    x = np.asarray(x, dtype=float)
    return np.where(np.isfinite(x) & (x > floor), x, np.nan)


def maybe_display(df):
    try:
        display(df)
    except Exception:
        print(df)


def pressure_axis(ax):
    ax.set_yscale("log")
    ax.invert_yaxis()
    ax.set_ylabel("Pressure [bar]")


def pressure_sort_top_to_bottom(Pbar, arr):
    Pbar = np.asarray(Pbar)
    arr = np.asarray(arr)
    idx = np.argsort(Pbar)

    if arr.shape[0] == len(Pbar):
        return Pbar[idx], arr[idx, ...]

    return Pbar[idx], arr


def get_from_dicts(key, *dicts, default=None):
    for d in dicts:
        if isinstance(d, dict) and key in d:
            return d[key]
    return default


def get_first_key(dicts, keys, default=None):
    for key in keys:
        for d in dicts:
            if isinstance(d, dict) and key in d:
                return d[key], key
    return default, None


def local_scale_height_from_pz(Pbar, z):
    Pbar = np.asarray(Pbar, dtype=float)
    z = np.asarray(z, dtype=float)

    order = np.argsort(z)
    z_sort = z[order]
    P_sort = Pbar[order]

    lnP = np.log(np.maximum(P_sort, 1e-300))
    dlnP_dz = np.gradient(lnP, z_sort)
    H_sort = 1.0 / np.maximum(np.abs(dlnP_dz), 1e-300)

    H = np.empty_like(H_sort)
    H[order] = H_sort
    return H


# ============================================================
# 3. Find and reuse old successful runs
# ============================================================

def find_all_catalogs(project_root):
    output_dir = project_root / "output"
    catalogs = []

    if output_dir.exists():
        for p in output_dir.rglob("*catalog*.csv"):
            if p.is_file():
                try:
                    df = pd.read_csv(p)
                    if "label" in df.columns and "status" in df.columns:
                        catalogs.append((p, df))
                except Exception:
                    pass

    return catalogs


def row_matches_case(row, case):
    try:
        same_flux = np.isclose(
            float(row.get("target_haze_flux")),
            float(case["target_haze_flux"]),
            rtol=1e-12,
            atol=0.0,
        )
        same_peak = np.isclose(
            float(row.get("gauss_peak_bar")),
            float(case["gauss_peak_bar"]),
            rtol=1e-12,
            atol=0.0,
        )
        same_sigma = np.isclose(
            float(row.get("gauss_sigma_dex")),
            float(case["gauss_sigma_dex"]),
            rtol=1e-12,
            atol=0.0,
        )
    except Exception:
        return False

    status = str(row.get("status", "")).lower()
    converged = str(row.get("converged", "True")).lower()

    result_npz = Path(str(row.get("result_npz", "")))

    good_file = result_npz.exists()
    good_status = status in ["success", "reference_success"]
    good_converged = converged in ["true", "1", "yes"]

    return same_flux and same_peak and same_sigma and good_status and good_converged and good_file


def find_existing_successful_row(case):
    catalogs = find_all_catalogs(PROJECT_ROOT)

    matches = []

    for catalog_path, df in catalogs:
        for _, row in df.iterrows():
            if row_matches_case(row, case):
                row_dict = row.to_dict()
                row_dict["source_catalog"] = str(catalog_path)
                matches.append(row_dict)

    if len(matches) == 0:
        return None

    # Prefer rows that already have direct-Mie tau files.
    def score(row):
        tau_csv = Path(str(row.get("tau_csv", "")))
        result_npz = Path(str(row.get("result_npz", "")))
        s = 0
        if tau_csv.exists():
            s += 10
        if result_npz.exists():
            s += 5
        if "formal_figures_flux_comparison_v2" in str(row.get("source_catalog", "")):
            s += 2
        return s

    matches = sorted(matches, key=score, reverse=True)
    return matches[0]


# ============================================================
# 4. Run missing cases or reuse existing cases
# ============================================================

old_OUTPUT_ROOT = globals().get("OUTPUT_ROOT", None)
old_RERUN_EXISTING = globals().get("RERUN_EXISTING", None)

# run_one_case uses the global OUTPUT_ROOT at call time.
OUTPUT_ROOT = RUN_ROOT
RERUN_EXISTING = False

catalog_rows = []
start_time = time.time()

for case in FLUX_LADDER_CASES:
    label = case["label"]

    print("\n" + "#" * 88)
    print("FLUX LADDER CASE:", label)
    print("#" * 88)

    reused = None

    if REUSE_EXISTING_SUCCESSFUL_CASES:
        reused = find_existing_successful_row(case)

    if reused is not None:
        print("Reusing existing successful run.")
        print("  source catalog:", reused.get("source_catalog"))
        print("  result:", reused.get("result_npz"))

        reused["status"] = "reused_success"
        reused["ladder_label"] = label
        reused["estimated_runtime_min"] = case["estimated_runtime_min"]
        catalog_rows.append(reused)

        pd.DataFrame(catalog_rows).to_csv(MASTER_CATALOG_PATH, index=False)
        continue

    if not RUN_NEW_MISSING_CASES:
        print("No existing run found and RUN_NEW_MISSING_CASES=False. Skipping.")
        skipped = dict(case)
        skipped["status"] = "skipped_missing"
        catalog_rows.append(skipped)
        pd.DataFrame(catalog_rows).to_csv(MASTER_CATALOG_PATH, index=False)
        continue

    elapsed = time.time() - start_time
    remaining = WALLTIME_SECONDS - elapsed
    estimate_s = case["estimated_runtime_min"] * 60.0

    print(f"elapsed [hr]: {elapsed/3600:.2f}")
    print(f"remaining [hr]: {remaining/3600:.2f}")
    print(f"estimated runtime [hr]: {estimate_s/3600:.2f}")

    if estimate_s > remaining * 0.90:
        print("Skipping: not enough walltime buffer to start this case.")
        skipped = dict(case)
        skipped["status"] = "skipped_walltime_buffer"
        skipped["elapsed_hr"] = elapsed / 3600.0
        skipped["remaining_hr"] = remaining / 3600.0
        catalog_rows.append(skipped)
        pd.DataFrame(catalog_rows).to_csv(MASTER_CATALOG_PATH, index=False)
        continue

    row = run_one_case(base_case, case)
    row["ladder_label"] = label
    row["estimated_runtime_min"] = case["estimated_runtime_min"]
    row["source_catalog"] = "new_ladder_run"
    catalog_rows.append(row)

    pd.DataFrame(catalog_rows).to_csv(MASTER_CATALOG_PATH, index=False)
    print("Master catalog saved:", MASTER_CATALOG_PATH)

# restore optional globals
if old_OUTPUT_ROOT is not None:
    OUTPUT_ROOT = old_OUTPUT_ROOT

if old_RERUN_EXISTING is not None:
    RERUN_EXISTING = old_RERUN_EXISTING

master_catalog = pd.DataFrame(catalog_rows)
master_catalog.to_csv(MASTER_CATALOG_PATH, index=False)

print("\nFlux ladder catalog:")
maybe_display(master_catalog)

success = master_catalog[
    master_catalog["status"].astype(str).str.lower().isin(
        ["success", "reused_success", "reference_success"]
    )
].copy()

if len(success) == 0:
    raise RuntimeError("No successful/reused runs available for figure production.")


# ============================================================
# 5. Ensure direct-Mie tau products exist
# ============================================================

def find_tau_csv_for_row(row):
    tau_csv = str(row.get("tau_csv", ""))

    if len(tau_csv) > 0 and Path(tau_csv).exists():
        return Path(tau_csv)

    run_dir = Path(str(row.get("run_dir", "")))

    if run_dir.exists():
        candidates = sorted(run_dir.rglob("*direct_mie_tau.csv"))
        if candidates:
            return candidates[0]

    return None


def ensure_tau_products(row):
    label = row["label"]

    tau_csv = find_tau_csv_for_row(row)

    if tau_csv is not None:
        return tau_csv

    if "compute_direct_mie_opacity" not in globals():
        print(f"{label}: no tau_csv and compute_direct_mie_opacity is unavailable.")
        return None

    result_npz = Path(str(row.get("result_npz", "")))

    if not result_npz.exists():
        print(f"{label}: no result_npz available for opacity recompute.")
        return None

    result = load_npz_as_dict(result_npz)
    run_dir = Path(str(row.get("run_dir", RUN_ROOT / label)))

    print(f"{label}: computing missing direct-Mie opacity products.")
    optics = compute_direct_mie_opacity(result, run_dir, label, n_wave=300)

    if optics is None:
        return None

    return Path(optics["summary"]["tau_csv"])


tau_paths = {}

for i, row in success.iterrows():
    tau_paths[row["label"]] = ensure_tau_products(row)

success["tau_csv_fixed"] = success["label"].map(lambda x: str(tau_paths.get(x, "")))
success.to_csv(MASTER_CATALOG_PATH, index=False)


# ============================================================
# 6. Load saved results and spectra
# ============================================================

results = {}
spectra = []

for _, row in success.iterrows():
    label = row["label"]
    result_npz = Path(str(row.get("result_npz", "")))

    if result_npz.exists():
        results[label] = load_npz_as_dict(result_npz)

    tau_csv = tau_paths.get(label, None)

    if tau_csv is not None and Path(tau_csv).exists():
        tau = pd.read_csv(tau_csv)
        tau["label"] = label
        tau["target_haze_flux"] = float(row["target_haze_flux"])

        if "transmission_strength" not in tau.columns and "slant_tau_ext" in tau.columns:
            tau["transmission_strength"] = 1.0 - np.exp(-tau["slant_tau_ext"])

        if "relative_slant_proxy" not in tau.columns and "slant_tau_ext" in tau.columns:
            tau_1um = np.interp(1.0, tau["wave_um"], tau["slant_tau_ext"])
            tau["relative_slant_proxy"] = np.log1p(tau["slant_tau_ext"]) - np.log1p(tau_1um)

        spectra.append(tau)

if len(results) == 0:
    raise RuntimeError("No result files loaded.")

spectra_all = pd.concat(spectra, ignore_index=True) if spectra else pd.DataFrame()

spectra_csv = TABLE_ROOT / "flux_ladder_spectra_long.csv"
if len(spectra_all) > 0:
    spectra_all.to_csv(spectra_csv, index=False)
    print("Saved spectra table:", spectra_csv)

print("Loaded result labels:", list(results.keys()))


# ============================================================
# 7. Derived microphysics diagnostics
# ============================================================

def derive_microphysics(result):
    Pbar = np.asarray(result["Pbar"], dtype=float)
    n = np.asarray(result["n"], dtype=float)
    m_bin = np.asarray(result["m_bin"], dtype=float)
    r_nm = np.asarray(result["r_compact_nm"], dtype=float)

    mass_bin = n * m_bin[None, :]
    mass_density = np.sum(mass_bin, axis=1)
    number_density = np.sum(n, axis=1)

    number_radius_nm = np.sum(n * r_nm[None, :], axis=1) / np.maximum(number_density, 1e-300)
    mass_radius_nm = np.sum(mass_bin * r_nm[None, :], axis=1) / np.maximum(mass_density, 1e-300)

    return {
        "Pbar": Pbar,
        "n": n,
        "m_bin": m_bin,
        "r_nm": r_nm,
        "mass_bin": mass_bin,
        "mass_density": mass_density,
        "number_density": number_density,
        "number_radius_nm": number_radius_nm,
        "mass_radius_nm": mass_radius_nm,
    }


derived = {label: derive_microphysics(result) for label, result in results.items()}


# ============================================================
# 8. Rebuild contexts and compute real timescales
# ============================================================

def rebuild_context_for_row(row):
    label = row["label"]

    print("\n" + "=" * 72)
    print("Rebuilding context for timescales:", label)
    print("=" * 72)

    reset_caligo_runtime_cache(base_case)

    attach_science_settings(
        base_case,
        target_haze_flux=float(row["target_haze_flux"]),
        gauss_peak_bar=float(row["gauss_peak_bar"]),
        gauss_sigma_dex=float(row["gauss_sigma_dex"]),
    )

    ensure_source_info_built(base_case)

    ctx = base_case.prepare_microphysics(
        climate_out=None,
        with_coagulation=True,
    )

    if isinstance(ctx, dict) and "ctx" in ctx and isinstance(ctx["ctx"], dict):
        ctx = ctx["ctx"]

    if not isinstance(ctx, dict):
        raise RuntimeError(f"{label}: context rebuild did not return a dictionary.")

    return ctx


def compute_timescales(label, result, ctx):
    Pbar = np.asarray(get_from_dicts("Pbar", result, ctx), dtype=float)
    n = np.asarray(get_from_dicts("n", result), dtype=float)
    m_bin = np.asarray(get_from_dicts("m_bin", result, ctx), dtype=float)
    r_nm = np.asarray(get_from_dicts("r_compact_nm", result, ctx), dtype=float)

    Kzz = get_from_dicts("Kzz", ctx, result)
    H = get_from_dicts("H", ctx, result)
    z = get_from_dicts("z", ctx, result)
    dz = get_from_dicts("dz", ctx, result)

    v_set, v_set_key = get_first_key(
        [ctx, result],
        ["v_set", "v_settle", "settling_velocity", "vfall", "v_fall"],
        default=None,
    )

    K_coag, K_coag_key = get_first_key(
        [ctx, result],
        ["K_coag", "coag_kernel", "Kij", "coagulation_kernel"],
        default=None,
    )

    thermal_rate, thermal_key = get_first_key(
        [ctx, result],
        [
            "thermal_rate",
            "k_therm",
            "k_thermal",
            "therm_destroy_rate",
            "thermal_destruction_rate",
            "k_loss_thermal",
        ],
        default=None,
    )

    if H is None:
        if z is not None:
            H = local_scale_height_from_pz(Pbar, z)
        elif dz is not None:
            H = np.ones_like(Pbar) * np.nanmedian(np.asarray(dz, dtype=float))
        else:
            raise RuntimeError(f"{label}: cannot compute H.")

    H = np.asarray(H, dtype=float)

    if Kzz is None:
        tau_mix = np.full_like(Pbar, np.nan)
    else:
        Kzz = np.asarray(Kzz, dtype=float)
        tau_mix = H**2 / np.maximum(Kzz, 1e-300)

    mass_bin = n * m_bin[None, :]
    mass_density = np.sum(mass_bin, axis=1)
    number_density = np.sum(n, axis=1)

    active_mask = mass_density > max(np.nanmax(mass_density) * 1e-30, 1e-300)

    tau_set_mass = np.full_like(Pbar, np.nan)
    tau_coag = np.full_like(Pbar, np.nan)
    tau_therm = np.full_like(Pbar, np.nan)

    if v_set is not None:
        v_set = np.asarray(v_set, dtype=float)

        if v_set.shape == n.shape:
            v_mass = np.sum(mass_bin * v_set, axis=1) / np.maximum(mass_density, 1e-300)
            tau_set_mass = H / np.maximum(v_mass, 1e-300)
            tau_set_mass[~active_mask] = np.nan

    if K_coag is not None:
        K_coag = np.asarray(K_coag, dtype=float)

        if K_coag.ndim == 3 and K_coag.shape[0] == n.shape[0]:
            freq_i = np.einsum("kij,kj->ki", K_coag, n)
            freq_eff = np.sum(n * freq_i, axis=1) / np.maximum(number_density, 1e-300)
            tau_coag = 1.0 / np.maximum(freq_eff, 1e-300)
            tau_coag[~active_mask] = np.nan

    if thermal_rate is not None:
        thermal_rate = np.asarray(thermal_rate, dtype=float)

        if thermal_rate.ndim == 1 and len(thermal_rate) == len(Pbar):
            tau_therm = 1.0 / np.maximum(thermal_rate, 1e-300)
            tau_therm[~active_mask] = np.nan

    df = pd.DataFrame({
        "label": label,
        "Pbar": Pbar,
        "tau_mix_s": tau_mix,
        "tau_set_mass_s": tau_set_mass,
        "tau_coag_s": tau_coag,
        "tau_therm_s": tau_therm,
        "mass_density_g_cm3": mass_density,
        "number_density_cm3": number_density,
        "mass_radius_nm": np.sum(mass_bin * r_nm[None, :], axis=1) / np.maximum(mass_density, 1e-300),
        "number_radius_nm": np.sum(n * r_nm[None, :], axis=1) / np.maximum(number_density, 1e-300),
    })

    print(f"{label}: timescale finite fractions")
    for col in ["tau_mix_s", "tau_set_mass_s", "tau_coag_s", "tau_therm_s"]:
        print(f"  {col}: {np.isfinite(df[col]).mean():.3f}")

    return df


contexts = {}
timescale_tables = []

for _, row in success.iterrows():
    label = row["label"]

    if label not in results:
        continue

    try:
        ctx = rebuild_context_for_row(row)
        contexts[label] = ctx

        tdf = compute_timescales(label, results[label], ctx)
        tdf["target_haze_flux"] = float(row["target_haze_flux"])

        timescale_tables.append(tdf)
    except Exception as exc:
        print(f"Timescale diagnostics failed for {label}: {repr(exc)}")
        print(traceback.format_exc())

timescales_all = pd.concat(timescale_tables, ignore_index=True)

timescale_csv = TABLE_ROOT / "flux_ladder_timescales_long.csv"
timescales_all.to_csv(timescale_csv, index=False)

print("Saved timescale table:", timescale_csv)


# ============================================================
# 9. Build enriched summary table
# ============================================================

summary_rows = []

for _, row in success.iterrows():
    label = row["label"]

    if label not in results:
        continue

    d = derived[label]
    tdf = timescales_all[timescales_all["label"] == label].copy()

    k_mass = int(np.nanargmax(d["mass_density"]))
    p_mass_peak = d["Pbar"][k_mass]

    if len(tdf) > 0:
        kt = int(np.nanargmin(np.abs(np.log10(tdf["Pbar"]) - np.log10(p_mass_peak))))
        tau_mix_peak = safe_float(tdf["tau_mix_s"].iloc[kt])
        tau_set_peak = safe_float(tdf["tau_set_mass_s"].iloc[kt])
        tau_coag_peak = safe_float(tdf["tau_coag_s"].iloc[kt])
        tau_therm_peak = safe_float(tdf["tau_therm_s"].iloc[kt])
    else:
        tau_mix_peak = tau_set_peak = tau_coag_peak = tau_therm_peak = np.nan

    srow = {
        "label": label,
        "target_haze_flux": safe_float(row.get("target_haze_flux")),
        "gauss_peak_bar": safe_float(row.get("gauss_peak_bar")),
        "column_mass_g_cm2": safe_float(row.get("column_mass_g_cm2")),
        "number_weighted_radius_nm_global": safe_float(row.get("number_weighted_radius_nm")),
        "mass_weighted_radius_nm_global": safe_float(row.get("mass_weighted_radius_nm")),
        "p_mass_peak_bar": safe_float(p_mass_peak),
        "mass_density_peak_g_cm3": safe_float(np.nanmax(d["mass_density"])),
        "mass_radius_at_peak_nm": safe_float(d["mass_radius_nm"][k_mass]),
        "number_radius_at_peak_nm": safe_float(d["number_radius_nm"][k_mass]),
        "tau_mix_at_mass_peak_s": tau_mix_peak,
        "tau_set_at_mass_peak_s": tau_set_peak,
        "tau_coag_at_mass_peak_s": tau_coag_peak,
        "tau_therm_at_mass_peak_s": tau_therm_peak,
        "max_vertical_tau_ext": safe_float(row.get("max_vertical_tau_ext")),
        "max_slant_tau_ext": safe_float(row.get("max_slant_tau_ext")),
        "max_transmission_strength": safe_float(row.get("max_transmission_strength")),
        "wave_um_at_max_tau": safe_float(row.get("wave_um_at_max_tau")),
        "p_eff_at_max_tau_bar": safe_float(row.get("p_eff_at_max_tau_bar")),
        "runtime_min": safe_float(row.get("runtime_min")),
        "result_npz": row.get("result_npz", ""),
        "tau_csv": str(tau_paths.get(label, "")),
    }

    summary_rows.append(srow)

summary = pd.DataFrame(summary_rows).sort_values("target_haze_flux")
summary_csv = TABLE_ROOT / "flux_ladder_summary_enriched.csv"
summary.to_csv(summary_csv, index=False)

print("\nEnriched summary:")
maybe_display(summary)

print("Saved summary:", summary_csv)


# ============================================================
# 10. Paper figure 1:
# Source flux -> column mass/radius/opacity summary
# ============================================================

fig, axes = plt.subplots(2, 2, figsize=(11.5, 8.5))
axes = axes.ravel()

sf = summary.sort_values("target_haze_flux")

ax = axes[0]
ax.plot(sf["target_haze_flux"], sf["column_mass_g_cm2"], marker="o", linewidth=2)
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel(r"Haze source flux [g cm$^{-2}$ s$^{-1}$]")
ax.set_ylabel(r"Haze column mass [g cm$^{-2}$]")
ax.set_title("Column haze mass")

ax = axes[1]
ax.plot(sf["target_haze_flux"], sf["mass_weighted_radius_nm_global"], marker="o", linewidth=2, label="global")
ax.plot(sf["target_haze_flux"], sf["mass_radius_at_peak_nm"], marker="s", linewidth=2, label="at mass peak")
ax.set_xscale("log")
ax.set_xlabel(r"Haze source flux [g cm$^{-2}$ s$^{-1}$]")
ax.set_ylabel("Mass-weighted radius [nm]")
ax.set_title("Particle size response")
ax.legend(fontsize=8)

ax = axes[2]
ax.plot(sf["target_haze_flux"], sf["max_slant_tau_ext"], marker="o", linewidth=2)
ax.axhline(1.0, linestyle="--", linewidth=1)
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel(r"Haze source flux [g cm$^{-2}$ s$^{-1}$]")
ax.set_ylabel(r"Max slant $\tau_{\rm ext}$")
ax.set_title("Transmission opacity")

ax = axes[3]
ax.plot(sf["target_haze_flux"], sf["max_transmission_strength"], marker="o", linewidth=2)
ax.set_xscale("log")
ax.set_xlabel(r"Haze source flux [g cm$^{-2}$ s$^{-1}$]")
ax.set_ylabel(r"Max $1-e^{-\tau_{\rm slant}}$")
ax.set_title("Approximate spectral impact")

fig.suptitle("Flux ladder: source strength to haze opacity", y=1.02, fontsize=15)
fig.tight_layout()

savefig(fig, FIG_ROOT / "figure_flux_ladder_summary.png")
plt.show()


# ============================================================
# 11. Paper figure 2:
# Source profile -> mass profile -> radius-pressure -> spectrum
# ============================================================

fig, axes = plt.subplots(2, 2, figsize=(12.0, 9.0))
axes = axes.ravel()

# A. source profile
ax = axes[0]
for label, result in results.items():
    if "S_mass_density" not in result:
        continue
    row = summary[summary["label"] == label].iloc[0]
    ax.plot(
        clean_positive(result["S_mass_density"]),
        result["Pbar"],
        linewidth=2,
        label=f"{row['target_haze_flux']:.1e}",
    )

ax.set_xscale("log")
pressure_axis(ax)
ax.set_xlabel(r"Source [g cm$^{-3}$ s$^{-1}$]")
ax.set_title("Haze source profile")
ax.legend(title=r"$F_{\rm haze}$", fontsize=8)

# B. final haze mass density
ax = axes[1]
for label, d in derived.items():
    row = summary[summary["label"] == label].iloc[0]
    ax.plot(
        clean_positive(d["mass_density"]),
        d["Pbar"],
        linewidth=2,
        label=f"{row['target_haze_flux']:.1e}",
    )

ax.set_xscale("log")
pressure_axis(ax)
ax.set_xlabel(r"Haze mass density [g cm$^{-3}$]")
ax.set_title("Final haze mass density")
ax.legend(title=r"$F_{\rm haze}$", fontsize=8)

# C. mass-weighted radius
ax = axes[2]
for label, d in derived.items():
    row = summary[summary["label"] == label].iloc[0]
    ax.plot(
        clean_positive(d["mass_radius_nm"]),
        d["Pbar"],
        linewidth=2,
        label=f"{row['target_haze_flux']:.1e}",
    )

ax.set_xscale("log")
pressure_axis(ax)
ax.set_xlabel("Mass-weighted radius [nm]")
ax.set_title("Particle growth with pressure")
ax.legend(title=r"$F_{\rm haze}$", fontsize=8)

# D. slant tau spectrum
ax = axes[3]
if len(spectra_all) > 0:
    for label in summary["label"]:
        df = spectra_all[spectra_all["label"] == label].copy()
        if len(df) == 0:
            continue
        row = summary[summary["label"] == label].iloc[0]
        ax.loglog(
            df["wave_um"],
            clean_positive(df["slant_tau_ext"]),
            linewidth=2,
            label=f"{row['target_haze_flux']:.1e}",
        )

ax.axhline(1.0, linestyle="--", linewidth=1)
ax.set_xlabel("Wavelength [µm]")
ax.set_ylabel(r"Slant $\tau_{\rm ext}$")
ax.set_title("Direct-Mie opacity")
ax.legend(title=r"$F_{\rm haze}$", fontsize=8)

fig.suptitle("Source flux to microphysics to opacity", y=1.02, fontsize=15)
fig.tight_layout()

savefig(fig, FIG_ROOT / "figure_source_to_microphysics_to_opacity.png")
plt.show()


# ============================================================
# 12. Paper figure 3:
# Timescale comparison at representative case and flux trend
# ============================================================

# choose highest-flux successful case as representative, because it has the strongest haze.
rep_label = summary.sort_values("target_haze_flux")["label"].iloc[-1]
rep = timescales_all[timescales_all["label"] == rep_label].copy()

fig, axes = plt.subplots(1, 2, figsize=(13.0, 5.3))

ax = axes[0]
curves = [
    ("tau_mix_s", r"mixing: $H^2/K_{zz}$"),
    ("tau_set_mass_s", r"settling: $H/v_{\rm set}$"),
    ("tau_coag_s", r"coagulation: $1/(Kn)$"),
    ("tau_therm_s", "thermal destruction"),
]

for col, lab in curves:
    if col not in rep.columns:
        continue
    y = clean_positive(rep[col])
    if np.isfinite(y).sum() == 0:
        continue
    ax.plot(y, rep["Pbar"], linewidth=2, label=lab)

for t_ref, name in [
    (3600.0, "1 hr"),
    (86400.0, "1 d"),
    (86400.0 * 30.0, "30 d"),
    (86400.0 * 365.0, "1 yr"),
]:
    ax.axvline(t_ref, linestyle=":", linewidth=0.8)
    ax.text(t_ref, np.nanmax(rep["Pbar"]), name, rotation=90, fontsize=8, va="bottom", ha="right")

ax.set_xscale("log")
pressure_axis(ax)
ax.set_xlabel("Timescale [s]")
ax.set_title(f"Timescales: {rep_label}")
ax.legend(fontsize=8)

ax = axes[1]
ax.plot(sf["target_haze_flux"], sf["tau_mix_at_mass_peak_s"], marker="o", linewidth=2, label="mixing")
ax.plot(sf["target_haze_flux"], sf["tau_set_at_mass_peak_s"], marker="o", linewidth=2, label="settling")
ax.plot(sf["target_haze_flux"], sf["tau_coag_at_mass_peak_s"], marker="o", linewidth=2, label="coagulation")
ax.plot(sf["target_haze_flux"], sf["tau_therm_at_mass_peak_s"], marker="o", linewidth=2, label="thermal")
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel(r"Haze source flux [g cm$^{-2}$ s$^{-1}$]")
ax.set_ylabel("Timescale at mass peak [s]")
ax.set_title("Timescale response at haze-mass peak")
ax.legend(fontsize=8)

fig.suptitle("Transport and microphysical timescales", y=1.03, fontsize=15)
fig.tight_layout()

savefig(fig, FIG_ROOT / "figure_timescales_representative_and_flux_trend.png")
plt.show()


# ============================================================
# 13. Paper figure 4:
# Pressure-radius mass maps for low/mid/high flux
# ============================================================

ordered_labels = list(summary.sort_values("target_haze_flux")["label"])

if len(ordered_labels) >= 3:
    map_labels = [ordered_labels[0], ordered_labels[len(ordered_labels)//2], ordered_labels[-1]]
else:
    map_labels = ordered_labels

fig, axes = plt.subplots(1, len(map_labels), figsize=(5.1 * len(map_labels), 4.6), sharey=True)

if len(map_labels) == 1:
    axes = [axes]

for ax, label in zip(axes, map_labels):
    d = derived[label]
    Pbar = d["Pbar"]
    r_nm = d["r_nm"]
    mass_bin = d["mass_bin"]

    Psort, mass_sort = pressure_sort_top_to_bottom(Pbar, mass_bin)

    vmax = np.nanmax(mass_sort)
    vmin = max(vmax * 1e-9, 1e-300)

    im = ax.pcolormesh(
        r_nm,
        Psort,
        np.maximum(mass_sort, vmin),
        shading="auto",
        norm=LogNorm(vmin=vmin, vmax=vmax),
    )

    ax.plot(d["mass_radius_nm"], Pbar, color="white", linewidth=2.0, label="mass-weighted")
    ax.plot(d["number_radius_nm"], Pbar, color="black", linewidth=1.5, label="number-weighted")

    row = summary[summary["label"] == label].iloc[0]

    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.invert_yaxis()
    ax.set_xlabel("Compact radius [nm]")
    ax.set_title(r"$F_{\rm haze}=$" + f"{row['target_haze_flux']:.1e}")

    if ax is axes[0]:
        ax.set_ylabel("Pressure [bar]")
        ax.legend(fontsize=7, loc="lower right")

fig.subplots_adjust(right=0.88, wspace=0.12)
cbar_ax = fig.add_axes([0.90, 0.15, 0.018, 0.70])
cb = fig.colorbar(im, cax=cbar_ax)
cb.set_label(r"Haze mass density by bin [g cm$^{-3}$]")

fig.suptitle("Flux dependence of the pressure-radius haze distribution", y=0.99, fontsize=15)

savefig(fig, FIG_ROOT / "figure_pressure_radius_mass_maps_flux_ladder.png")
plt.show()


# ============================================================
# 14. Paper figure 5:
# Spectral opacity comparison
# ============================================================

if len(spectra_all) > 0:
    fig, axes = plt.subplots(1, 2, figsize=(13.0, 5.2))

    ax = axes[0]
    for label in summary["label"]:
        df = spectra_all[spectra_all["label"] == label].copy()
        if len(df) == 0:
            continue
        row = summary[summary["label"] == label].iloc[0]

        ax.loglog(
            df["wave_um"],
            clean_positive(df["slant_tau_ext"]),
            linewidth=2,
            label=f"{row['target_haze_flux']:.1e}",
        )

    ax.axhline(1.0, linestyle="--", linewidth=1)
    ax.set_xlabel("Wavelength [µm]")
    ax.set_ylabel(r"Direct-Mie slant $\tau_{\rm ext}$")
    ax.set_title("Slant haze optical depth")
    ax.legend(title=r"$F_{\rm haze}$", fontsize=8)

    ax = axes[1]
    for label in summary["label"]:
        df = spectra_all[spectra_all["label"] == label].copy()
        if len(df) == 0:
            continue
        row = summary[summary["label"] == label].iloc[0]

        ax.semilogx(
            df["wave_um"],
            df["transmission_strength"],
            linewidth=2,
            label=f"{row['target_haze_flux']:.1e}",
        )

    ax.set_xlabel("Wavelength [µm]")
    ax.set_ylabel(r"$1-e^{-\tau_{\rm slant}}$")
    ax.set_title("Approximate transmission impact")
    ax.legend(title=r"$F_{\rm haze}$", fontsize=8)

    fig.suptitle("Flux dependence of haze spectral opacity", y=1.03, fontsize=15)
    fig.tight_layout()

    savefig(fig, FIG_ROOT / "figure_direct_mie_spectral_comparison_flux_ladder.png")
    plt.show()


# ============================================================
# 15. Figure 6:
# Source-flux scaling diagnostics
# ============================================================

fig, axes = plt.subplots(1, 3, figsize=(15.0, 4.7))

ax = axes[0]
ax.plot(sf["target_haze_flux"], sf["column_mass_g_cm2"], marker="o", linewidth=2)
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel(r"$F_{\rm haze}$ [g cm$^{-2}$ s$^{-1}$]")
ax.set_ylabel(r"Column mass [g cm$^{-2}$]")
ax.set_title("Mass scaling")

# Add reference linear scaling anchored at first point.
if len(sf) >= 2:
    x0 = sf["target_haze_flux"].iloc[0]
    y0 = sf["column_mass_g_cm2"].iloc[0]
    xline = np.array([sf["target_haze_flux"].min(), sf["target_haze_flux"].max()])
    yline = y0 * (xline / x0)
    ax.plot(xline, yline, linestyle="--", linewidth=1, label="linear")
    ax.legend(fontsize=8)

ax = axes[1]
ax.plot(sf["target_haze_flux"], sf["mass_weighted_radius_nm_global"], marker="o", linewidth=2)
ax.set_xscale("log")
ax.set_xlabel(r"$F_{\rm haze}$ [g cm$^{-2}$ s$^{-1}$]")
ax.set_ylabel("Global mass-weighted radius [nm]")
ax.set_title("Growth response")

ax = axes[2]
ax.plot(sf["target_haze_flux"], sf["max_slant_tau_ext"], marker="o", linewidth=2, label=r"$\tau_{\rm slant,max}$")
ax.plot(sf["target_haze_flux"], sf["max_transmission_strength"], marker="s", linewidth=2, label=r"$1-e^{-\tau}$")
ax.axhline(1.0, linestyle="--", linewidth=1)
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel(r"$F_{\rm haze}$ [g cm$^{-2}$ s$^{-1}$]")
ax.set_ylabel("Opacity diagnostic")
ax.set_title("Opacity response")
ax.legend(fontsize=8)

fig.suptitle("Nonlinear response to haze source flux", y=1.03, fontsize=15)
fig.tight_layout()

savefig(fig, FIG_ROOT / "figure_source_flux_scaling_diagnostics.png")
plt.show()


# ============================================================
# 16. Manifest
# ============================================================

manifest = {
    "project_root": str(PROJECT_ROOT),
    "ladder_root": str(LADDER_ROOT),
    "walltime_hours": WALLTIME_HOURS,
    "cases": FLUX_LADDER_CASES,
    "master_catalog": str(MASTER_CATALOG_PATH),
    "summary_csv": str(summary_csv),
    "timescale_csv": str(timescale_csv),
    "spectra_csv": str(spectra_csv) if len(spectra_all) > 0 else "",
    "fig_root": str(FIG_ROOT),
    "figure_files": [str(p) for p in sorted(FIG_ROOT.glob("*.png"))],
}

manifest_path = LADDER_ROOT / "manifest.json"

with open(manifest_path, "w") as f:
    json.dump(manifest, f, indent=2)

print("\nDONE.")
print("Paper figure output directory:")
print(FIG_ROOT)
print("\nSummary table:")
print(summary_csv)
print("\nManifest:")
print(manifest_path)

print("\nFigures created:")
for p in sorted(FIG_ROOT.glob("*.png")):
    print(" ", p.name)
```
